```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class BSTIterator(object):
    # This represents an iterator over the in-order traversal of a binary search tree (BST)
    # Inorder: Left -> Root -> Right

    def __init__(self, root):
        """
        :type root: Optional[TreeNode]
        """

        # The pointer should be initialized to
        # a non-existent number smaller than any element in the BST.

        self.tree_list = []
        self.createList(root)
        self.pointer = -1


    def createList(self, root):
        # Base case
        if not root:
            return

        self.createList(root.left)

        # We append a value here because it's inorder.
        self.tree_list.append(root.val)

        self.createList(root.right)



    def next(self):
        """
        :rtype: int
        """

        # You may assume that next() calls will always be valid.
        # That is, there will be at least a next number in the in-order traversal
        # when next() is called.

        self.pointer += 1
        return self.tree_list[self.pointer]


    def hasNext(self):
        """
        :rtype: bool
        """

        # Returns true if there exists a number in the traversal
        # to the right of the pointer, otherwise returns false.
        if self.pointer + 1 < len(self.tree_list):
            return True

        return False
```
