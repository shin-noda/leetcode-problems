```python
class Solution(object):
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """

        result = []

        def backtrack(current_string, open_count, closed_count):
            # Base caase: the string is complete
            if len(current_string) == 2 * n:
                result.append(current_string)
                return

            # Can we add an opening brackt?
            if open_count < n:
                backtrack(current_string + "(", open_count + 1, closed_count)

            # Can we add a closing bracket?
            if closed_count < open_count:
                backtrack(current_string + ")", open_count, closed_count + 1)

        backtrack("", 0, 0)

        return result
```

```python
solution = Solution()
```

```python
n = 3

print(solution.generateParenthesis(n))
```

```python
n = 1

print(solution.generateParenthesis(n))
```
