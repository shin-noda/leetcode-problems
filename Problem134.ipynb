```python
class Solution(object):
    def canCompleteCircuit(self, gas, cost):
        """
        :type gas: List[int]
        :type cost: List[int]
        :rtype: int
        """

        """
        We apply the logic behind Raney's Lemma:
        First, we ensure that the total gas available is at least equal
        to the total cost (sum(gas) >= sum(cost)).

        If a solution exists, it is guaranteed to be unique.
        By tracking the cumulative net gas as we 'drive' through the array,
        the starting point must be the station immediately following the
        global minimum (the lowest point in our gas balance).
        """

        net = 0
        starting_index = -1

        # Initialize min with infinity so the first station always updates it
        min_gas_balance = float('inf')
        lowest_point_index = 0

        for x in range(len(gas)):
            net = net + gas[x] - cost[x]

            if net < min_gas_balance:
                min_gas_balance = net
                lowest_point_index = x

        # If the total net gas is negative, the trip is impossible
        if net < 0:
            starting_index = -1

        else:
            # The starting point is the station after the lowest point.
            # We use modulo (%) to circle back to 0 if the lowest point was the last index.
            starting_index = (lowest_point_index + 1) % len(gas)

        return starting_index
```

```python
solution = Solution()
```

```python
gas = [1,2,3,4,5]
cost = [3,4,5,1,2]

print(solution.canCompleteCircuit(gas, cost))
```

```python
gas = [2,3,4]
cost = [3,4,3]

print(solution.canCompleteCircuit(gas, cost))
```

```python
gas = [0]
cost = [0]

print(solution.canCompleteCircuit(gas, cost))
```
