```python
# Definition for a Node.
class Node(object):
    def __init__(self, val=0, left=None, right=None, next=None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
```

```python
class Solution(object):
    def connect(self, root):
        """
        :type root: Node
        :rtype: Node
        """

        # The first node of the current level
        curr = root

        # Each loop represent one level and
        # the loop continues when there is a level below.
        while curr:
            dummy = Node(0)

            # tail moves forward as we add nodes
            tail = dummy

            # Walk across the current level node by node.
            # We build the next level's .next chain.
            while curr:
                if curr.left:
                    tail.next = curr.left
                    tail = tail.next

                if curr.right:
                    tail.next = curr.right
                    tail = tail.next

                curr = curr.next

            # dummy.next points to the first node of the next level
            curr = dummy.next

        return root
```
