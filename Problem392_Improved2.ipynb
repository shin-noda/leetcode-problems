```python
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        """
        We preprocess the string into
        an inverted index of character positions,
        then use binary search to greedily find the next valid position
        for each character in the subsequence.

        The time complexity: O(|t| + |s| * log|t|)

        The Pre-processing Phase: O(|t|)
            This is for building the map.
            |t| -> The length of t, len(t)


        The Query Phase: O(|s| * log|t|)
            Iterate through s: |s| steps

            Binary Search: On a list of size N takes O(log N)
                For the entire string t: O(log|t|)
                For eavery character in s, one single check takes O(|s| * log|t|)
        """

        # 1. Pre-process t: Store indices for each character
        # Using a list of 26 lists for lowercase English letters
        pos_map = [[] for _ in range(26)]
        for i, char in enumerate(t):
            pos_map[ord(char) - ord('a')].append(i)

        # This helper finds the first element in 'arr' strictly greater than 'target'
        def binary_search(arr, target):
            low = 0
            high = len(arr) - 1
            res = -1
            while low <= high:
                # The // operator: the integer division
                mid = (low + high) // 2
                if arr[mid] > target:
                    res = arr[mid]
                    high = mid - 1

                else:
                    low = mid + 1

            return res

        # 2. Process s
        curr_pos = -1
        for char in s:
            indices = pos_map[ord(char) - ord('a')]
            # Find the next available index for this character after curr_pos
            next_pos = binary_search(indices, curr_pos)

            if next_pos == -1:
                return False

            curr_pos = next_pos

        return True
```

```python
solution = Solution()
```
