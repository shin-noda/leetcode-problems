```python
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """

        result = []
        carry = 0

        # Pointers for the end of both strings
        i = len(a) - 1
        j = len(b) - 1

        # Process until both strings are exhausted and no carry remains
        while i >= 0 or j >= 0 or carry:
            # Get the current digits (0 if string is exhausted)
            d_a = 0
            d_b = 0

            if i >= 0:
                d_a = int(a[i])

            if j >= 0:
                d_b = int(b[j])

            # Calculate sum and carry
            current_sum = d_a + d_b + carry
            result.append(str(current_sum % 2))
            carry = current_sum // 2

            i -= 1
            j -= 1

        # Join and reverse the list
        return "".join(result[::-1])
```

```python
solution = Solution()
```

```python
a = "11"
b = "1"

print(solution.addBinary(a, b))
```
