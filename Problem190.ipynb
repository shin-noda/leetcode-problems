```python
class Solution(object):
    def reverseBits(self, n):
        """
        :type n: int
        :rtype: int
        """

        if n == 0:
            return 0

        # Convert a number to a binary number
        b_n = []
        while n > 0:
            r = int(n % 2)
            b_n.append(r)
            n = n // 2

        for i in range(len(b_n), 32):
            b_n.append(0)

        result = 0
        for i in range(len(b_n)):
            result += pow(2, i) * b_n[len(b_n) - 1 - i]

        return result
```

```python
solution = Solution()
```
