```python
class Solution(object):
    def myPow(self, x, n):
        """
        :type x: float
        :type n: int
        :rtype: float
        """

        # Handle negative n
        if n < 0:
            x = 1 / x
            n = -n

        result = 1.0
        current_product = x

        while n > 0:
            # If n is odd, multiply the result by the current power
            if n % 2 == 1:
                result *= current_product

            # Square the base and halve the exponent
            current_product *= current_product
            n = n // 2

        return result
```

```python
solution = Solution()
```

```python
x = 2.00000
n = 10

print(solution.myPow(x, n))
```
