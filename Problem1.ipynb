```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """

        # Note: only one valid answer exists

        # {'num':[index1, index2, index3, ...]}
        # e.g. If 3 appears twice in nums at 0 and 2, {'3':[0, 2]}
        num_map = {}
        result = []

        for i in range(len(nums)):
            num = nums[i]
            if num not in num_map:
                # Add the iandex of the nums for num
                num_map[num] = [i]

            else:
                # Add the index to the list in the map
                num_map[num].append(i)

        # Go through the map
        for num in num_map:
            val = target - num

            # The first condition is for the case like
            # target: 6, nums = [3, 3]
            if num == val and len(num_map[num]) == 2:
                # Get the first and the second indexed numbers
                result.append(num_map.get(num)[0])
                result.append(num_map.get(num)[1])
                break

            elif num != val and val in num_map:
                result.append(num_map.get(num)[0])
                result.append(num_map.get(val)[0])
                break

        return result
```

```python
solution = Solution()
```

```python
nums = [2,7,11,15]
target = 9

print(solution.twoSum(nums, target))
```

```python
nums = [3,2,4]
target = 6

print(solution.twoSum(nums, target))
```

```python
nums = [3,3]
target = 6

print(solution.twoSum(nums, target))
```
