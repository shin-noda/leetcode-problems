```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """

        if x < 0:
            return False

        digit = 0
        tmp = x
        while tmp > 0:
            tmp = tmp // 10
            digit += 1

        while x > 0:
            last = x % 10
            first = x // pow(10, digit - 1)

            if first != last:
                return False

            # We remove the first and the last digits
            x = (x - last - (first * pow(10, digit - 1))) // 10

            # We need to decrement by 2 since we remove the first and the last digit
            digit -= 2

        return True
```

```python
solution = Solution()
```
