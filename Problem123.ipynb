```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """

        # We calculate the profit from the first and the second trades first and
        # combine two results later.
        first_profit = self.calculateMaxProfit(prices, 0, len(prices))
        second_profit = self.calculateProfitToTheRight(prices, 0, len(prices))

        max_val = 0
        for fp, sp in zip(first_profit, second_profit):
            max_val = max(max_val, fp + sp)

        return max_val


    def calculateMaxProfit(self, prices, start, end):
        max_profit = 0
        record = []

        # Check if this is the last day
        if start >= len(prices):
            return 0

        max_val = prices[start]
        min_val = prices[start]

        for i in range(start, end):
            # If the current price is lower than the min value,
            # it updates both min and max values.
            if min_val > prices[i]:
                min_val = prices[i]
                max_val = prices[i]

            else:
                max_val = max(max_val, prices[i])
                max_profit = max(max_profit, max_val - min_val)

            # Keep the record
            record.append(max_profit)

        return record


    def calculateProfitToTheRight(self, prices, start, end):
        max_profit = 0
        record = []

        max_val = prices[end - 1]

        for i in range(end - 1, -1, -1):
            max_val = max(max_val, prices[i])
            max_profit = max(max_profit, max_val - prices[i])

            # Keep the record
            record.insert(0, max_profit)

        return record
```

```python
solution = Solution()
```
