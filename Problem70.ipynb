```python
class Solution(object):
    def climbStairs(self, n):
        """
        :type n: int
        :rtype: int
        """

        steps = 0

        if n == 0:
            return 1

        if n == 1:
            return 1

        n0 = 1
        n1 = 1

        for i in range(2, n + 1):
            steps = n0 + n1
            n0 = n1
            n1 = steps

        return steps
```

```python
solution = Solution()
```

```python
n = 2

print(solution.climbStairs(n))
```

```python
n = 3

print(solution.climbStairs(n))
```
