```python
class Solution(object):
    def insert(self, intervals, newInterval):
        """
        :type intervals: List[List[int]]
        :type newInterval: List[int]
        :rtype: List[List[int]]
        """

        # Note that you don't need to modify intervals in-place.
        # You can make a new array and return it.

        # intervals: non-overlapping and sorted in ascending order by start_i

        results = []
        i = 0
        n = len(intervals)

        new_s, new_e = newInterval

        # 1. Add all intervals that come strictly before the new interval
        while i < n and intervals[i][1] < new_s:
            results.append(intervals[i])
            i += 1

        # 2. Merge all overlapping intervals
        # An overlap exists if the interval start is <= the new interval's end
        while i < n and intervals[i][0] <= new_e:
            new_s = min(new_s, intervals[i][0])
            new_e = max(new_e, intervals[i][1])
            i += 1

        # Add the merged "mega-interval"
        results.append([new_s, new_e])

        # 3. Add the rest of the intervals that come after
        while i < n:
            results.append(intervals[i])
            i += 1

        return results
```
