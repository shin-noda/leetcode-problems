```python
class Solution(object):
    def findPeakElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # We use binary search to find the local maxima.

        left = 0
        right = len(nums) - 1

        while left < right:
            mid = (left + right) // 2

            # If we are on an upward slope, the peak is to the right.
            if nums[mid] < nums[mid + 1]:
                left = mid + 1

            # If we are on the down slope, the peak is to the left.
            else:
                right = mid

        return left
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]

print(solution.findPeakElement(nums))
```

```python
nums = [1,2,1,3,5,6,4]

print(solution.findPeakElement(nums))
```
