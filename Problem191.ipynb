```python
class Solution(object):
    def hammingWeight(self, n):
        """
        :type n: int
        :rtype: int
        """

        # Set Bit: A set bit refers to a bit in the binary representation
        #  of a number that has a value of 1.

        set_bit = 0

        while n > 0:
            r = (n % 2)

            if r == 1:
                set_bit += 1

            n = n // 2

        return set_bit
```

```python
solution = Solution()
```

```python
n = 11

print(solution.hammingWeight(n))
```
