```python
from typing import List
```

```python
# Brute Force approach
class Brute_Solution:
    def productExceptSelf(self, nums):
        # Create an array of 1
        answer = [1] * len(nums)

        for i in range(len(nums)):
            element = nums[i]
            for j in range(len(nums)):
                if i == j:
                    answer[j] = answer[j] * 1

                else:
                    answer[j] = answer[j] * element

        return answer

obj = Brute_Solution()

nums1 = [1,2,3,4]
print(obj.productExceptSelf(nums1))

nums2 = [-1,1,0,-3,3]
print(obj.productExceptSelf(nums2))
```

```python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums)
        answer = [1] * n

        # prefix product
        prefix = 1
        for i in range(n):
            answer[i] = prefix
            prefix *= nums[i]

        # suffix product
        suffix = 1
        for i in range(n-1, -1, -1):
            answer[i] *= suffix
            suffix *= nums[i]

        return answer

obj = Solution()

nums1 = [1,2,3,4]
print(obj.productExceptSelf(nums1))

nums2 = [-1,1,0,-3,3]
print(obj.productExceptSelf(nums2))
```
