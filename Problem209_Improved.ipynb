```python
class Solution(object):
    def minSubArrayLen(self, target, nums):
        """
        :type target: int
        :type nums: List[int]
        :rtype: int
        """

        """
        This uses the combination of prefix sums and binary search.
        """

        n = len(nums)

        # Build prefix sum array
        # It's n + 1 because it calculates from index 0
        # e.g. nums = [2, 3, 1, 2, 4, 3]
        # prefix = [0, 2, 5, 6, 8, 12, 15]
        prefix = [0] * (n + 1)
        for i in range(n):
            prefix[i + 1] = prefix[i] + nums[i]

        # This finds the rightmost index j such that prefix[j] <= needed
        # We find the largest j satisfying that giving us
        # the shortest subarray ending at i - 1
        def binary_search_rightmost(prefix, needed):
            left = 0
            right = len(prefix) - 1

            # default invalid index
            index = -1

            # This is <= instead of < because
            # there is a case that len(nums) = 1.
            while left <= right:
                mid = (left + right) // 2

                # The min prefix[mid] = 0 at prefix[0]
                # so if needed is negative, index is always -1.
                if prefix[mid] <= needed:
                    index = mid
                    left = mid + 1
                else:
                    right = mid - 1

            return index

        min_len = float('inf')

        # For each prefix sum, use binary search to find the valid subarray
        for i in range(1, n + 1):
            # We want prefix[j] <= prefix[i] - target
            needed = prefix[i] - target

            # It returns the rightmost index where prefix[j] >= needed + 1
            # so we subtract 1 to get prefix[j] <= needed
            j = binary_search_rightmost(prefix, needed)
            if j != -1:
                min_len = min(min_len, i - j)

        result = 0

        if min_len != float('inf'):
            result = min_len

        return result
```

```python
solution = Solution()
```

```python
target = 7
nums = [2,3,1,2,4,3]

print(solution.minSubArrayLen(target, nums))
```
