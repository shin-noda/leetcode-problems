```python
class Solution(object):
    def gameOfLife(self, board):
        """
        :type board: List[List[int]]
        :rtype: None Do not return anything, modify board in-place instead.
        """

        """
        Rules:
        1. Any live cell with fewer than two live neighbors dies
           as if caused by under-population.
        2. Any live cell with two or three live neighbors lives on to
           the next generation.
        3. Any live cell with more than three live neighbors dies,
           as if by over-population.
        4. Any dead cell with exactly three live neighbors becomes a live cell,
           as if by reproduction.
        """

        # Vertical Length (number of rows)
        m = len(board)

        # Horizontal Length (number of columns)
        n = len(board[0])

        # We can solve this in space by introducing new two states
        # which preserve the previous states.
        # Dead
        D = 0

        # Alive
        L = 1

        # Dead but Live next
        D_L = 2

        # Lived but Die next
        L_D = 3

        # We check the board and update the state using D_L and L_D
        for i in range(m):
            for j in range(n):
                # Number of neighbours
                nn = 0
                cell = board[i][j]

                # Check 8 direction
                # Top-Left
                if i - 1 >= 0 and j - 1 >= 0:
                    tmp = board[i - 1][j - 1]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Top
                if i - 1 >= 0:
                    tmp = board[i - 1][j]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Top-Right
                if i - 1 >= 0 and j + 1 <= n - 1:
                    tmp = board[i - 1][j + 1]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Left
                if j - 1 >= 0:
                    tmp = board[i][j - 1]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Right
                if j + 1 <= n - 1:
                    tmp = board[i][j + 1]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Bottom-Left
                if i + 1 <= m - 1 and j - 1 >= 0:
                    tmp = board[i + 1][j - 1]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Bottom
                if i + 1 <= m - 1:
                    tmp = board[i + 1][j]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Bottom-Right
                if i + 1 <= m - 1 and j + 1 <= n - 1:
                    tmp = board[i + 1][j + 1]

                    if tmp == L or tmp == L_D:
                        nn += 1

                # Check the number of neighbours
                # and update the board accordingly
                if nn < 2:
                    if cell == L:
                        board[i][j] = L_D

                # We skip the case when nn == 2
                elif nn == 3:
                    if cell == D:
                        board[i][j] = D_L

                elif nn > 3:
                    if cell == L:
                        board[i][j] = L_D

        # We make the board only 0's and 1's
        for i in range(m):
            for j in range(n):
                cell = board[i][j]

                if cell == D_L:
                    board[i][j] = L

                elif cell == L_D:
                    board[i][j] = D

        # print(board)
```

```python
solution = Solution()
```
