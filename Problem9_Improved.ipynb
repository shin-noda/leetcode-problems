```python
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """

        # 1. Negative numbers are never palindromes
        # 2. Numbers ending in 0 aren't palindromes (except 0 itself)
        if x < 0 or (x % 10 == 0 and x != 0):
            return False

        reverted_num = 0
        x_backup = x

        while x > 0:
            # Get the last digit
            last_digit = x % 10

            # Shift the reverted number left and add the new digit
            reverted_num = (reverted_num * 10) + last_digit

            # Remove the last digit from x
            x = x // 10

        return x_backup == reverted_num
```

```python
solution = Solution()
```
