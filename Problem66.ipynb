```python
class Solution(object):
    def plusOne(self, digits):
        """
        :type digits: List[int]
        :rtype: List[int]
        """

        carry = 0
        i = 1
        pointer = len(digits) - 1
        while True:
            num = digits[pointer]

            # This is only for the first iteration.
            if i == 1:
                num = num + 1

            # Add carry
            num = num + carry

            digits[pointer] = num % 10
            carry = num // 10

            if carry == 0:
                break

            i += 1
            pointer -= 1

            if len(digits) < i:
                digits.insert(0, 0)

                # Increment the pointer
                pointer += 1

        return digits
```

```python
solution = Solution()
```

```python
digits = [1,2,3]

print(solution.plusOne(digits))
```
