```python
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """

        # This is for storing words in a nested way like
        # [["bat"],["nat","tan"],["ate","eat","tea"]]
        anagram_list = []

        # This stores date like {'0':{'a': 1},'1':{'a':2,'b':3}}
        index_map_map = {}

        for word in strs:
            # For a word 'bat', it's stored like {'b':1,'a':1,'t':1}
            word_map = {}

            for s in word:
                word_map[s] = word_map.get(s, 0) + 1

            is_stored = False
            for index in index_map_map:
                if word_map == index_map_map.get(index):
                    anagram_list[index].append(word)
                    is_stored = True

            # It's a new item
            if not is_stored:
                index_map_map[len(index_map_map)] = word_map
                anagram_list.append([word])

        return anagram_list
```

```python
solution = Solution()
```

```python
strs = ["eat","tea","tan","ate","nat","bat"]

print(solution.groupAnagrams(strs))
```

```python
strs = [""]

print(solution.groupAnagrams(strs))
```
