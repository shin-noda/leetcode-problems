```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # Base case: if it's one node, it's already sorted.
        if not head or not head.next:
            return head

        # Split the list (the sabotage)
        mid = self.getMid(head)

        tmp = mid.next
        mid.next = None

        # Recursively sort (the breadcrumbs)
        # This keeps diving until it hits the base case
        left_side = self.sortList(head)
        right_side = self.sortList(tmp)

        # Merget them back together
        # This function returns the new head of the merged section
        # Note: this merge may be called multiple times during execution, not once.
        return self.merge(left_side, right_side)


    def getMid(self, head):
        slow = head
        fast = head.next

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

        return slow


    def merge(self, left, right):
        dummy = ListNode()

        # This is my moving pointer
        head = dummy

        while left and right:
            if left.val < right.val:
                head.next = left
                left = left.next

            else:
                head.next = right
                right = right.next

            # advance the head
            head = head.next

        # The clean up
        if left:
            head.next = left

        else:
            head.next = right

        return dummy.next
```
