```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        # Check if the root node is empty
        if not root:
            return 0

        count = self.checkDepth(root, 1)

        return count

    # This method checks if the node exists and
    # when it does, it returns 1. Otherwise, it returns 1
    def checkDepth(self, node, count):
        left = node.left
        right = node.right

        c1 = count
        c2 = count
        if left:
            c1 = self.checkDepth(left, count + 1)

        if right:
            c2 = self.checkDepth(right, count + 1)

        return max(c1, c2)
```
