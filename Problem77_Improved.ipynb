```python
class Solution(object):
    def combine(self, n, k):
        """
        :type n: int
        :type k: int
        :rtype: List[List[int]]
        """

        # We implement the method that performs nCk, n choose k.

        results = []
        combo = []

        def backtrack(start):
            # Base Case
            if len(combo) == k:
                results.append(combo[:])
                return

            for i in range(start, n + 1):
                # Choose
                combo.append(i)

                # Explore
                backtrack(i + 1)

                # Un-choose (Backtrack)
                combo.pop()

        backtrack(1)

        return results
```

```python
solution = Solution()
```

```python
n = 4
k = 2

print(solution.combine(n, k))
```
