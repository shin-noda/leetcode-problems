```python
class Solution(object):
    def findKthLargest(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """

        # This is a brute force solution: O(n log n)
        nums.sort()

        return nums[-k]
```

```python
solution = Solution()
```

```python
nums = [3,2,1,5,6,4]
k = 2

print(solution.findKthLargest(nums, k))
```

```python
nums = [3,2,3,1,2,4,5,5,6]
k = 4

print(solution.findKthLargest(nums, k))
```
