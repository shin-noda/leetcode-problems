```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        if not list1 and not list2:
            return None

        elif not list1:
            return list2

        elif not list2:
            return list1

        dummy = ListNode()
        current = dummy

        while list1 or list2:
            val1 = 0
            val2 = 0

            if list1 and list2:
                val1 = list1.val
                val2 = list2.val

                if val1 == val2:
                    current.next = ListNode(val1)
                    current = current.next

                    current.next = ListNode(val2)
                    current = current.next

                    list1 = list1.next
                    list2 = list2.next

                elif val1 < val2:
                    current.next = ListNode(val1)
                    current = current.next

                    list1 = list1.next

                elif val1 > val2:
                    current.next = ListNode(val2)
                    current = current.next

                    list2 = list2.next

            elif list1 and not list2:
                val1 = list1.val
                current.next = ListNode(val1)
                current = current.next

                list1 = list1.next

            elif list2 and not list1:
                val2 = list2.val
                current.next = ListNode(val2)
                current = current.next

                list2 = list2.next

        return dummy.next
```

```python
solution = Solution()
```
