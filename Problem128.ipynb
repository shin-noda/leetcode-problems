```python
class Solution(object):
    def longestConsecutive(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Check if nums is empty
        result = 0
        if len(nums) == 0:
            return result

        # I need to write an algorithm that runs in O(n) time => no sort()
        num_map = {}

        for num in nums:
            # We don't care if the numbers in nums are duplicated or not
            num_map[num] = 0

        for num in nums:
            counter = 0
            tmp = num

            # We decrement the value
            while True:
                if tmp in num_map:
                    if num_map[tmp] == 1:
                        # Already checked
                        break

                    elif num_map[tmp] == 0:
                        num_map[tmp] = 1
                        counter += 1
                        tmp -= 1

                    else:
                        break

                else:
                    break

            tmp = num + 1
            # We increment the value
            while True:
                if tmp in num_map:
                    if num_map[tmp] == 1:
                        # Already checked
                        break

                    elif num_map[tmp] == 0:
                        num_map[tmp] = 1
                        counter += 1
                        tmp += 1

                else:
                    break

            if counter > result:
                result = counter

        return result
```

```python
solution = Solution()
```
