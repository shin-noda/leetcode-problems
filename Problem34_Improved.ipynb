```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """

        # Constraint: Runtime complexity must be O(log n)
        # Use a binary search

        # Helper function to find the leftmost boundary
        def findBound(isFirst):
            left = 0
            right = len(nums) - 1
            bound = -1

            while left <= right:
                mid = (left + right) // 2

                if nums[mid] == target:
                    bound = mid

                    if isFirst:
                        # Keep looking left to find the very first index
                        right = mid - 1

                    else:
                        # Keep looking right to find the very last index
                        left = mid + 1

                elif nums[mid] < target:
                    left = mid + 1

                else:
                    right = mid - 1

            return bound

        start = findBound(True)

        # If we didn't find the start, the target isn't there at all
        if start == -1:
            return [-1, -1]

        end = findBound(False)

        return [start, end]
```

```python
solution = Solution()
```
