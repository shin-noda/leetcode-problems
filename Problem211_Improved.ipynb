```python
class TrieNode(object):
    def __init__(self):
        self.children = {}
        self.is_end = False


class WordDictionary(object):
    def __init__(self):
        self.root = TrieNode()


    def addWord(self, word):
        """
        :type word: str
        :rtype: None
        """

        node = self.root
        for c in word:
            if c not in node.children:
                node.children[c] = TrieNode()

            node = node.children[c]

        node.is_end = True


    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """

        # We use a helper function for the recursive DFS
        def dfs(j, root):
            curr = root

            for i in range(j, len(word)):
                char = word[i]

                if char == ".":
                    # For a dot, we check every existing child branch
                    for child in curr.children.values():
                        if dfs(i + 1, child):
                            return True

                    # If none of the branches worked, this path is dead
                    return False

                else:
                    # Standard Trie traversal
                    if char not in curr.children:
                        return False

                    curr = curr.children[char]

            return curr.is_end

        return dfs(0, self.root)
```
