```python
class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """

        dp = [0] * (len(s) + 1)

        # We mark the first element as 1
        dp[0] = 1
        for i in range(len(s)):
            if dp[i] == 0:
                continue

            for word in wordDict:
                if i + len(word) < len(dp) and word == s[i:i + len(word)]:
                    # Mark the index after the end of the word as 1
                    dp[i + len(word)] = 1

        return dp[-1] == 1
```

```python
solution = Solution()
```

```python
s = "leetcode"
wordDict = ["leet","code"]

print(solution.wordBreak(s, wordDict))
```
