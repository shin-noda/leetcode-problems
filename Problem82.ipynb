```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # Check if it's an empty node
        if not head:
            return head

        # A sentinel node
        dummy = ListNode(float('inf'))
        dummy.next = head

        curr = dummy
        prev = dummy

        # Intentionally put negative to so that
        # it meets the condition for the first iteration
        unique_val = -curr.val
        while curr:
            new_val = curr.val

            if (
                unique_val != new_val and
                not curr.next or
                curr.next and new_val != curr.next.val
            ):

                # We update the unique_val
                unique_val = new_val

                # New backup
                prev = curr

            # The value is not uniuque, so we proceed until
            # the value is not new_val
            else:
                # We remove the node
                prev.next = curr.next
                while curr.next and curr.next.val == new_val:
                    curr = curr.next
                    prev.next = curr.next

            curr = curr.next

        return dummy.next
```
