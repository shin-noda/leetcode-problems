```python
class ListNode(object):
    def __init__(self, key=0, next=None, prev=None):
        self.key = key
        self.next = next
        self.prev = prev
```

```python
class LRUCache(object):
    def __init__(self, capacity):
        """
        :type capacity: int
        """

        self.capacity = capacity

        # We create two nodes.
        # They stay as head and tails node which means
        # when a new value is created, it's inserted
        # between head and tail node.
        # the real head node is head.prev and the reals last node is
        # tail.next.
        self.head = ListNode(0)
        self.tail = ListNode(0)

        self.head.prev = self.tail
        self.tail.next = self.head

        self.val_map = {}
        self.add_map = {}


    def get(self, key):
        """
        :type key: int
        :rtype: int
        """

        if key in self.val_map:
            self.move_to_top(key)

            return self.val_map[key]

        else:
            return -1


    def put(self, key, value):
        """
        :type key: int
        :type value: int
        :rtype: None
        """

        if key in self.val_map:
            # It updates the value in the LRU cache and
            # the node goes to the top

            # Update the value
            self.val_map[key] = value

            self.move_to_top(key)
        else:
            if len(self.val_map) + 1 > self.capacity:
                # We need to evict a LRU node
                evicted_node = self.tail.next
                lru_key = evicted_node.key
                next = evicted_node.next

                self.tail.next = next
                next.prev = self.tail

                evicted_node.next = None
                evicted_node.prev = None

                # Remove from maps
                self.val_map.pop(lru_key)
                self.add_map.pop(lru_key)

            # Add new node
            new_node = ListNode(key)

            self.val_map[key] = value
            self.add_map[key] = new_node

            head_before = self.head.prev

            new_node.next = self.head
            new_node.prev = head_before

            self.head.prev = new_node
            head_before.next = new_node


    def move_to_top(self, key):
        curr = self.add_map.get(key)
        prev = curr.prev
        next = curr.next

        # Connect the node before and after curr
        prev.next = next
        next.prev = prev

        # Connect curr to the head and the head before
        head_before = self.head.prev
        curr.next = self.head
        curr.prev = head_before
        head_before.next = curr
        self.head.prev = curr
```
