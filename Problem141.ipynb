```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """

        # Check if head is empty
        if not head:
            return False

        map = {}

        val = head.val
        next = head.next
        while next != None:
            if next not in map:
                map[next] = 0

            else:
                return True

            val = next.val
            next = next.next

        return False
```
