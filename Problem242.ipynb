```python
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        # Create two hashmaps (dicts) to compare
        # For each map, we count the frequency of characters
        s_map = {}
        t_map = {}

        # t is not an anagram of s
        if len(s) != len(t):
            return False

        for a, b in zip(s, t):
            s_map[a] = s_map.get(a, 0) + 1
            t_map[b] = t_map.get(b, 0) + 1

        if s_map != t_map:
            return False

        return True
```

```python
solution = Solution()
```

```python
s = "anagram"
t = "nagaram"

print(solution.isAnagram(s, t))
```
