```python
class Solution(object):
    def isHappy(self, n):
        """
        :type n: int
        :rtype: bool
        """

        num_map = {}
        num = n

        # It breaks the loop when num is 1 or num is in num_map
        while True:
            tmp = 0
            s_num = str(num)

            for digit in s_num:
                tmp += pow(int(digit), 2)

            num = tmp

            # Break the loop
            if num == 1 or num in num_map:
                break

            # We need to add a num after the if above
            num_map[num] = 0

        # n is not a happy number
        if num != 1:
            return False

        return True
```

```python
solution = Solution()
```

```python
n = 19

print(solution.isHappy(n))
```

```python
n = 2

print(solution.isHappy(n))
```
