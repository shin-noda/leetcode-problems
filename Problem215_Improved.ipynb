```python
class Solution(object):
    def findKthLargest(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """

        # Time Complexity: O(n log k)
        # Space Complexity: O(k)

        # Build a min-heap using the fisrt k elements
        # Valid indices: 0, 1, 2, ..., k - 1
        # The rule of min-heap: parent <= children
        # Note: we don't check if children >= parent.
        # It's mathematically the same but we only check if
        # the node is smaller than its children nodes.
        # We always fix the parent node.
        # The chidren doens't go up. The parent go down.
        heap = nums[:k]

        # Function to maintain min-heap property (sift down)
        # Fix the node at index idx so that the subtree rooted at idx
        # becomes a valid min-heap.
        # Assumption: the left and right subtrees are already valid heaps.
        # Time Complexity: O(log k)
        def min_heapify(idx, heap_size):
            # We assume the current node is the smallest amont itself and its children.
            # We'll try to prove that wrong.
            smallest = idx

            # Find children.
            left = 2 * idx + 1
            right = 2 * idx + 2

            if left < heap_size and heap[left] < heap[smallest]:
                smallest = left

            if right < heap_size and heap[right] < heap[smallest]:
                smallest = right

            if smallest != idx:
                # We move the smaller value up, and push the bigger value down
                heap[idx], heap[smallest] = heap[smallest], heap[idx]

                # We move down because after swapping, the value we pushed down
                # might now violate the heap rule further down.
                min_heapify(smallest, heap_size)

        # Left child exists if: 2 * i + 1 < k
        # Flip the condition.
        # Left child doesn't exist if: 2 * i + 1 >= k
        # In the loop below, (k // 2) - 1 is the
        # last node that has chidren (last non-leaf)

        # So when is a node a leaf?
        # We already established that
        # 2 * i + 1 >= k -> leaf
        # We want to know from which index onward are all nodes leaves?
        # We had 2 * i + 1 >= k. Solve it for i.
        # 2 * i >= k - 1
        # i >= (k - 1) // 2
        # Since - 1 // 2 = 0 in Python integer math:
        # i >= k // 2.
        # Thus, all nodes starting at index k // 2 are leaves
        # Hence, the last parent node is k // 2 - 1.

        # Note: nodes after the last parent are already valid,
        # so we start the loop from the last parent node.

        # Build initial heap: start from the last non-leaf node and go up
        for i in range((k // 2) - 1, -1, -1):
            min_heapify(i, k)

        # At this point, heap is a min-heap of size k, and it stores
        # the k largetst elements we've seen so far.
        # Note: heap[0] is the smallest among those k largest.

        # Process the remaining elements
        for i in range(k, len(nums)):
            # heap[0] is the smallest among heap.
            if nums[i] > heap[0]:
                # Replace the smallest of the top-k
                heap[0] = nums[i]

                # Sift it down
                min_heapify(0, k)

        # The root is now the kth largets
        return heap[0]
```

```python
solution = Solution()
```

```python
nums = [3,2,1,5,6,4]
k = 2

print(solution.findKthLargest(nums, k))
```
