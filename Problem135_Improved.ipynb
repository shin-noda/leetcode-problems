```python
class Solution(object):
    def candy(self, ratings):
        """
        :type ratings: List[int]
        :rtype: int
        """

        """
        Use the peak valley method.
        We consider 3 situations:
            - Up
            - Down
            - Flat

        when we walk through the array.
        """

        total_candies = 1
        n = len(ratings)
        up = 0
        down = 0
        peak = 0

        for i in range(1, n):
            if ratings[i] > ratings[i - 1]:
                # Going up
                up = up + 1
                down = 0
                peak = up

                # +1 for adding the base candy
                total_candies = total_candies + up + 1

            elif ratings[i - 1] == ratings[i]:
                # Going flat
                up = 0
                down = 0
                peak = 0

                total_candies = total_candies + 1

            else:
                # Going down
                up = 0
                down = down + 1

                if down > peak:
                    # The downward slope has exceeded the peak's height.
                    # We add 1 candy to everyone in the 'down' sequence,
                    # plus an extra 1 to 'stretch' the peak upward.
                    total_candies = total_candies + down + 1
                else:
                    # We add 1 candy to everyone currently in the 'down' sequence.
                    total_candies = total_candies + down

        return total_candies
```

```python
solution = Solution()
```
