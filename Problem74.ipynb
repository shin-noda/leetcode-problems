```python
class Solution(object):
    def searchMatrix(self, matrix, target):
        """
        :type matrix: List[List[int]]
        :type target: int
        :rtype: bool
        """

        row = 0
        left = 0
        right = len(matrix) - 1

        # First, we seach the row that the target potentially exists
        while left <= right:
            row = (left + right) // 2

            if matrix[row][0] == target:
                return True

            elif matrix[row][0] < target:
                left = row + 1

            else:
                right = row - 1

        # We found the row
        # Note: if the target value is less that matrix[0][0], row is -1.
        row = left - 1

        if row < 0:
            return False

        # Finally, we find the value in the row
        left = 0
        right = len(matrix[row]) - 1
        while left <= right:
            mid = (left + right) // 2

            if matrix[row][mid] == target:
                return True

            elif matrix[row][mid] < target:
                left = mid + 1

            else:
                right = mid - 1

        return False
```

```python
solution = Solution()
```

```python
matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 3

print(solution.searchMatrix(matrix, target))
```
