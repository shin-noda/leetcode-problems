```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def averageOfLevels(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[float]
        """

        avg_list = []
        if not root:
            return avg_list

        queue = [root]

        while queue:
            # s: sum.
            s = float(0.0)
            avg = float(0.0)
            n = len(queue)

            for _ in range(n):
                node = queue.pop(0)

                s += float(node.val)

                if node.left:
                    queue.append(node.left)

                if node.right:
                    queue.append(node.right)

            avg = s / float(n)
            avg_list.append(avg)

        return avg_list
```
