```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class BSTIterator(object):
    # This represents an iterator over the in-order traversal of a binary search tree (BST)
    # Inorder: Left -> Root -> Right

    def __init__(self, root):
        """
        :type root: Optional[TreeNode]
        """

        # The pointer should be initialized to
        # a non-existent number smaller than any element in the BST.

        self.stack = []
        self._push_all_left(root)


    def _push_all_left(self, node):
        while node:
            self.stack.append(node)
            node = node.left


    def next(self):
        """
        :rtype: int
        """

        # You may assume that next() calls will always be valid.
        # That is, there will be at least a next number in the in-order traversal
        # when next() is called.

        # Pop the top (the next smallest element)
        node = self.stack.pop()

        # If there's a right child, we need to find the
        # smallest element in that right subtree
        if node.right:
            self._push_all_left(node.right)

        return node.val


    def hasNext(self):
        """
        :rtype: bool
        """

        # Returns true if there exists a number in the traversal
        # to the right of the pointer, otherwise returns false.

        # If the stack has nodes, there's a "next" element
        return len(self.stack) > 0
```
