```python
class Solution(object):
    def rotate(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """

        # Note: This algorithm has to be in-place => O(1) extra space

        n = len(matrix)

        # We move the starting point diagonally
        for i in range(n // 2):
            for j in range(n - 2 * i - 1):
                # From tl to tr. Moves at the top
                top = matrix[i][i + j]

                # From tr to br. Moves at the right
                right = matrix[i + j][n - i - 1]

                # From br to bl. Moves at the bottom
                bottom = matrix[n - i - 1][n - i - 1 - j]

                # From bl to tl. Moves at the left
                left = matrix[n - 1 - i - j][i]

                # Assign top to right
                matrix[i + j][n - i - 1] = top

                # Assign right to bottom
                matrix[n - i - 1][n - i - 1 - j] = right

                # Assign bottom to left
                matrix[n - 1 - i - j][i] = bottom

                # Assign left to top
                matrix[i][i + j] = left

        # print(matrix)
```

```python
solution = Solution()
```

```python
matrix = [
    [1,2,3]
    ,[4,5,6]
    ,[7,8,9]
]

print(solution.rotate(matrix))
```
