```python
class Solution(object):
    def myPow(self, x, n):
        """
        :type x: float
        :type n: int
        :rtype: float
        """

        if n < 0:
            x = 1 / x
            n = -n

        # x^0 = 1 where x != 0
        if n == 0:
            return 1

        if x == 1.0:
            return 1.0

        if x == -1.0:
            if n % 2 == 0:
                return 1.0

            else:
                return -1.0

        i = 1
        unit = x
        while i < n:
            x *= unit

            if x == 0.0:
                return 0.0

            i += 1

        return x
```

```python
solution = Solution()
```

```python
x = 2.00000
n = 10

print(solution.myPow(x, n))
```
