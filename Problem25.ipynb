```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def reverseKGroup(self, head, k):
        """
        :type head: Optional[ListNode]
        :type k: int
        :rtype: Optional[ListNode]
        """

        # Helper function which performs reversal
        def node_reversal(left, right, dummy):
            prev = dummy

            # Move the prev before left.
            for _ in range(left - 1):
                prev = prev.next

                # Check if prev exists
                if not prev:
                    return False

            # 'test' is just for checking if the 'right' pointer exits
            # This is prev.next, not prev because
            # prev is moved just before 'left', so prev.next is the left pointer
            # and to move test to 'right', it needs to move k steps
            # from prev.next, not prev.

            # prev.next is the location of the 'left' pointer.
            test = prev.next

            # Make sure left exists.
            if not test:
                return False

            # Move the test pointer to the right
            # to make sure that it exists
            for _ in range(right - left):
                test = test.next

                # Check if test exists
                if not test:
                    return False

            # Set up curr
            curr = prev.next

            # Peform reversal
            for _ in range(right - left):
                tmp = curr.next
                curr.next = tmp.next
                tmp.next = prev.next
                prev.next = tmp

            return True

        # No swiching needed
        if k <= 1:
            return head

        # A sentinel node
        dummy = ListNode(0)
        dummy.next = head

        is_going = True
        left = 1
        right = left + (k - 1)
        while is_going:
            # Perform reversal
            is_going = node_reversal(left, right, dummy)

            # Shift left and right by k
            left += k
            right += k

        return dummy.next
```
