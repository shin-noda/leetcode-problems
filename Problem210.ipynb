```python
class Solution(object):
    def findOrder(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: List[int]
        """

        # Use Kahn's Algorithm

        # Build an adjacency list
        # e.g. adj[0] = [1, 2] means course 0 unlocks course 1 and 2.
        adj = []
        for _ in range(numCourses):
            adj.append([])

        # e.g. indegree[3] = 2 means course 3 needs two other courses first.
        indegree = [0] * numCourses

        # Count how many prerequisite each course has.
        for dest, src in prerequisites:
            # src: a prerequisite, dest: a course
            # src must be done before dest.
            adj[src].append(dest)

            # Increment a number of prerequisites that the course has to take first.
            indegree[dest] += 1

        # Find "free courses" that you don't need prerequisites
        queue = []
        for i in range(len(indegree)):
            if indegree[i]== 0:
                queue.append(i)

        # Manual pointer. We don't use pop().
        head = 0

        # It continues while we take courses.
        # Note: number of courses only increase, not decrease since we don't pop().
        # This means len(queue) only increase and we can return queue in the end.
        while head < len(queue):
            # Course we can take
            # Note: we don't pop() but use a manual pointer, head.
            course = queue[head]

            head += 1

            # Unlock dependent courses
            for neighbor in adj[course]:
                # Decrement a number of dependent courses
                indegree[neighbor] -= 1

                # Check if a course becomes available
                if indegree[neighbor] == 0:
                    # We can take it now
                    queue.append(neighbor)

        # Check if we took the number of courses that were required.
        if len(queue) != numCourses:
            return []

        return queue
```

```python
solution = Solution()
```
