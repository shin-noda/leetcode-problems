```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """

        perms = []
        perms_set = set()
        results = []

        def backtracking(vals):
            # Base case
            if len(perms) == len(nums):
                # Note: have to use the slicing
                # If we use perms, it will not produce the right results
                results.append(perms[:])
                return

            for i in range(len(vals)):
                num = vals[i]
                temp = vals[:]

                perms.append(num)
                temp.pop(i)

                backtracking(temp)

                # Backtrack
                perms.pop()

        vals = nums[:]
        backtracking(vals)

        return results
```

```python
solution = Solution()
```

```python
nums = [1,2,3]

print(solution.permute(nums))
```
