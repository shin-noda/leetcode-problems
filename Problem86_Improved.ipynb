```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def partition(self, head, x):
        """
        :type head: Optional[ListNode]
        :type x: int
        :rtype: Optional[ListNode]
        """

        # Create two dummy heads for our two separate lists
        before_dummy = ListNode(0)
        after_dummy = ListNode(0)

        # Pointers to the current last node of the two lists
        # We need both before_dummy and before (and after_dummy and after)
        # so that before_dummy won't be altered later whereas
        # before will be modified.
        before = before_dummy
        after = after_dummy

        # We don't "need" curr but we preserve the head.
        curr = head
        while curr:
            if curr.val < x:
                # Attach to the 'before' list
                before.next = curr
                before = before.next
            else:
                # Attach to the 'after' list
                after.next = curr
                after = after.next

            curr = curr.next

        # Crucial: Terminate the 'after' list to avoid cycles
        after.next = None

        # Stitch the two lists together
        before.next = after_dummy.next

        return before_dummy.next
```
