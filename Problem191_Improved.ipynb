```python
class Solution(object):
    def hammingWeight(self, n):
        """
        :type n: int
        :rtype: int
        """

        # Set Bit: A set bit refers to a bit in the binary representation
        #  of a number that has a value of 1.

        # Use Brian Kernighan's Algorithm

        # e.g. n = 12
        # n = 12:     1 1 0 0
        # n - 1 = 11: 1 0 1 1

        # Now, look at them column by column using the AND rule (only 1 & 1 stays 1):
        # Leftmost bit: 1 & 1 = 1
        # Second bit:   1 & 0 = 0 <-- The rightmost original 1 is now dead!
        # Third bit:    0 & 1 = 0
        # Fourth bit:   0 & 1 = 0

        # Result: 1 0 0 0

        set_bit = 0

        while n > 0:
            # Clear the least significant set bit
            n &= (n - 1)
            set_bit += 1

        return set_bit
```

```python
solution = Solution()
```
