```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def partition(self, head, x):
        """
        :type head: Optional[ListNode]
        :type x: int
        :rtype: Optional[ListNode]
        """

        # Check if nodes are empty
        if not head:
            return head

        # The sentinel node
        dummy = ListNode(0, head)

        # The rightmost node to the left of the head.
        left = dummy

        # The previous node
        prev = dummy

        # Number of left nodes
        nl = 0
        i = 1
        while head:
            if x > head.val:
                nl += 1

                # If the number of left nodes is the same as i,
                # we don't have to switch nodes
                if nl != i:
                    # To perfor a switch, we need to do the following things.
                    # 1. make a copy of the current head
                    # 2. make a copy of head.next
                    # 3. connect prev.next to head.next
                    # 4. connect head.next to left.next
                    # 5. connect left.next to head
                    # 6. update left to head
                    # 7. update head to tmp
                    # Note: since we remove the head, no need to update prev.

                    # Step 1.
                    tmp = head

                    # Step 2.
                    next = head.next

                    # Step 3.
                    prev.next = head.next

                    # Step 4.
                    head.next = left.next

                    # Step 5.
                    left.next = head

                    # Step 6.
                    left = head

                    # Step 7.
                    head = next

                else:
                    # Update the left node
                    left = head

                    prev = head
                    head = head.next

            else:
                prev = head
                head = head.next

            i += 1

        return dummy.next
```
