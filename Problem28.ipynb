```python
class Solution(object):
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """

        n_h = len(haystack)
        n_n = len(needle)

        # Check if strings are empty or needle is longer than haystack
        if n_h == 0 or n_n == 0 or n_n > n_h:
            return -1

        # Check if all the characters in needle are in haystack
        unique_haystack_set = set(haystack)
        unique_needle_set = set(needle)

        is_needle_subset = unique_needle_set.issubset(unique_haystack_set)

        if not is_needle_subset:
            return -1

        result = -1
        index = 0

        # the first letter of needle
        f_n = needle[0]
        while index + n_n <= n_h:
            tmp = 0
            if f_n == haystack[index]:
                while tmp < n_n:
                    if haystack[index + tmp] == needle[tmp]:
                        tmp = tmp + 1

                    else:
                        break

                if tmp >= n_n:
                    result = index
                    break

            index = index + 1

        return result
```

```python
solution = Solution()
```

```python
haystack = "sadbutsad"
needle = "sad"

print(solution.strStr(haystack, needle))
```
