```python
class Solution(object):
    def findMedianSortedArrays(self, nums1, nums2):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :rtype: float
        """

        # If the total length (nums1 + nums2) is odd → return the middle element
        # If the total length is even → return the average of the two middle elements

        # Ensure nums1 is the smaller array to keep time complexity O(log(min(m, n)))
        if len(nums1) > len(nums2):
            # Swap so that nums1 is always the smaller array
            # We only do binary search on nums1.
            nums1, nums2 = nums2, nums1

        m, n = len(nums1), len(nums2)

        # Binary search range on nums1
        low, high = 0, m

        # Number of elements that should be on the left side in total
        # We are splitting it so that the LEFT side contains the median elment
        # when the total length is odd
        half_len = (m + n + 1) // 2

        # We are splitting both arrays into:
        # LEFT SIDE | RIGHT SIDE
        #
        # The goal:
        # Every element in LEFT <= every element in RIGHT
        #
        # Because arrays are sorted, we only need to check boundary elements:
        #
        # max(LEFT) = max(left1, left2)
        # min(RIGHT) = min(right1, right2)
        #
        # So the correct partition condition is:
        # max(left1, left2) <= min(right1, right2)
        #
        # Which is equivalent to:
        # left1 <= right2 AND left2 <= right1

        while low <= high:
            # i = number of elements taken from nums1 into the LEFT side
            # j = number of elements taken from nums2 into the LEFT side
            # We force: i + j = half_len

            i = (low + high) // 2
            j = half_len - i

            # Initialize boundary values
            # We use infinity so edge cases work naturally
            left1 = float('-inf')
            right1 = float('inf')
            left2 = float('-inf')
            right2 = float('inf')

            # nums1 split: [ ... left1 | right1 ... ]
            if i > 0:
                # Last element on the left side of nums1
                left1 = nums1[i - 1]

            if i < m:
                # First element on the right side of nums1
                right1 = nums1[i]

            # nums2 split: [ ... left2 | right2 ... ]
            if j > 0:
                # Last element on the left side of nums2
                left2 = nums2[j - 1]

            if j < n:
                # First element on the right side of nums2
                right2 = nums2[j]

            # ------------------------------------------------------------
            # Check if we found the correct partition
            #
            # Meaning:
            # Nothing on the left side should be greater than anything on the right side
            # ------------------------------------------------------------
            if left1 <= right2 and left2 <= right1:

                # If total length is odd → median is the max of left side
                if (m + n) % 2 == 1:
                    # The median is the last element of LEFT
                    return float(max(left1, left2))

                # If total length is even → median is average of middle two values
                return (max(left1, left2) + min(right1, right2)) / 2.0

            # ------------------------------------------------------------
            # If partition is not correct, adjust binary search
            # ------------------------------------------------------------
            # We took too many elements from nums1 → move left
            elif left1 > right2:
                high = i - 1

            # We took too few elements from nums1 → move right
            else:
                low = i + 1
```
