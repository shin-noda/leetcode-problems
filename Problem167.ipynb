```python
class Solution(object):
    def twoSum(self, numbers, target):
        """
        :type numbers: List[int]
        :type target: int
        :rtype: List[int]
        """

        # solution must use only constant extra space.
        # -> O(1) space
        # -> This is not allowed: indices = [0] * n  # n is input size
        # Ask: “Does the extra space I use change if the input gets bigger?”
        # Yes → Not constant
        # No → Constant

        n = len(numbers)
        left = 0
        right = n - 1

        while left < right:
            sum = numbers[left] + numbers[right]

            if sum == target:
                break

            elif sum > target:
                right = right - 1

            else:
                left = left + 1

        # We increment 1 for both indices
        return [left + 1, right + 1]
```

```python
solution = Solution()
```

```python
numbers = [2,7,11,15]
target = 9

print(solution.twoSum(numbers, target))
```

```python
numbers = [2,3,4]
target = 6

print(solution.twoSum(numbers, target))
```
