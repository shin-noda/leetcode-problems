```python
class TrieNode:
    def __init__(self):
        # Using a fixed-size list or a dict is fine but
        # dict is more meoory efficient for sparse trees
        self.children = {}
        self.is_end = False


class Solution(object):
    def wordBreak(self, s, wordDict):
        """
        :type s: str
        :type wordDict: List[str]
        :rtype: bool
        """

        # Build the Trie from scratch
        root = TrieNode()
        for word in wordDict:
            curr = root

            for char in word:
                if char not in curr.children:
                    curr.children[char] = TrieNode()

                curr = curr.children[char]

            curr.is_end = True

        # DP Array (Boolean)
        n = len(s)
        dp = [False] * (n + 1)
        dp[0] = True

        # Traverse the string
        for i in range(n):
            if not dp[i]:
                continue

            # From here, see what words we can find in the Trie
            curr = root
            for j in range(i, n):
                char = s[j]

                if char not in curr.children:
                    break

                curr = curr.children[char]
                if curr.is_end:
                    dp[j + 1] = True

        return dp[n]
```

```python
solution = Solution()
```

```python
s = "leetcode"
wordDict = ["leet","code"]

print(solution.wordBreak(s, wordDict))
```
