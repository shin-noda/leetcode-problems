```python
class Solution(object):
    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """

        # The num should be between 1 and 3999
        if not (num >= 1 and num <= 3999):
            return -1

        roman_map = {
            1: 'I', 5: 'V', 10: 'X', 50: 'L', 100: 'C', 500: 'D', 1000: 'M'
        }

        # Check if the number is within the roman_map
        if num in roman_map:
            return roman_map[num]

        roman_numeral = ""
        str_num = str(num)
        n = len(str_num)

        for i in range(len(str_num)):
            # This digits is either 1, 10, 100, or 1000
            digits = pow(10, (n - 1 - i))
            tmp_num = int(str_num[i]) * digits
            single_num = int(tmp_num / digits)

            # When single_num is 0, we do nothing.
            # So, it's not in this block.
            if 1 <= single_num <= 3:
                for x in range(single_num):
                    roman_numeral = roman_numeral+ roman_map[digits]

            elif single_num == 4:
                # Use the subtractive form
                roman_numeral = roman_numeral + roman_map[digits] + roman_map[5 * digits]

            elif 5 <= single_num <= 8:
                # Add a roman numeral corresponding to 5, 50, or 500
                roman_numeral = roman_numeral + roman_map[5 * digits]

                # This is for the case single_num is 6, 7, or 8
                for j in range(single_num - 5):
                    roman_numeral = roman_numeral + roman_map[digits]

            elif single_num == 9:
                # Use the subtractive form
                roman_numeral = roman_numeral + roman_map[digits] + roman_map[digits * 10]

        return roman_numeral
```

```python
solution = Solution()
```
