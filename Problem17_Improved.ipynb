```python
class Solution(object):
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """

        # Time Complexity: O(n * 4^n) where n is len(digit) and 4 is the at most
        # number of digits
        # We generate 4^n combinations and each takes O(n) to construct

        # Space Complexity: O(n * 4^n)
        # Auxiliary Space: O(n) -> Recursion depth, n.

        # Idea: Use Backtracking.
        # Backtracking is memory-efficient because it reuses space (DFS), while
        # Cartesian expansion stores whole layers (BFS).

        # Check if this is empty
        if not digits:
            return []

        # Idea: Build a Cartesian Product

        # Note: no '0's and '1's
        digit_map = {
            "2": "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "pqrs",
            "8": "tuv",
            "9": "wxyz"
        }

        results = []

        # Use an inner method
        def backtrack(index, path):
            # Base case
            if index >= len(digits):
                results.append(path)
                return

            letters = digit_map[digits[index]]
            for c in letters:
                backtrack(index + 1, path + c)

        backtrack(0, "")

        return results
```

```python
solution = Solution()
```

```python
digits = "23"

print(solution.letterCombinations(digits))
```

```python
digits = "2"

print(solution.letterCombinations(digits))
```

```python
digits = "234"

print(solution.letterCombinations(digits))
```
