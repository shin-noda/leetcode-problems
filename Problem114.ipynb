```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def flatten(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: None Do not return anything, modify root in-place instead.
        """

        # Check if root is empty.
        if not root:
            return None

        dummy = TreeNode()
        dummy.next = root

        self.add_list = []
        self.add_list.append(root)

        self.helper(root)

        for i in range(len(self.add_list) - 1):
            node1 = self.add_list[i]
            node2 = self.add_list[i + 1]

            node1.left = None
            node2.left = None

            node1.right = node2


    def helper(self, root):
        left = root.left
        right = root.right

        if left:
            self.add_list.append(left)
            self.helper(left)

        if right:
            self.add_list.append(right)
            self.helper(right)
```
