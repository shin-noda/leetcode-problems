```python
class Solution(object):
    def findMinArrowShots(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """

        # Check if points are empty
        if not points:
            return 0

        # Sort by the end coordinate (index 1)
        # O(N log N)
        # Sort the points list, but judge each point based on its second value.
        # Another examples:
        # e.g. sort by the length of strings: list.sort(key=lambda x: len(x))
        # e.g. sort by the absolute value of x: list.sort(key=lambda x: abs(x))
        points.sort(key=lambda x: x[1])

        arrows = 1
        # Place the first arrow at the end of the first balloon
        current_arrow_pos = points[0][1]

        for i in range(1, len(points)):
            # If the current balloon starts AFTER the last arrow position,
            # the old arrow can't hit it.
            if points[i][0] > current_arrow_pos:
                arrows += 1
                # Update the arrow position to the end of this balloon
                current_arrow_pos = points[i][1]

        return arrows
```

```python
solution = Solution()
```
