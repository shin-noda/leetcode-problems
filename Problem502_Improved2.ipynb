```python
import heapq

class Solution(object):
    def findMaximizedCapital(self, k, w, profits, capital):
        """
        :type k: int
        :type w: int
        :type profits: List[int]
        :type capital: List[int]
        :rtype: int
        """

        # Pair capital and profits together
        # O(N) space, O(N log N) to sort by required capital ascending
        projects = sorted(zip(capital, profits), key=lambda x: x[0])

        max_profit_heap = []
        pointer = 0
        n = len(projects)

        # Main greedy loop: we run at most k times
        for _ in range(k):
            # Push all projects we can currently affort into our max-heap
            while pointer < n and projects[pointer][0] <= w:
                # Store profit as a negative number to simulate a Max-Heap
                heapq.heappush(max_profit_heap, -projects[pointer][1])
                pointer += 1

            # If we can't afford any new projects and our heap is empty, we're stuck
            if not max_profit_heap:
                break

            # Take the project with the highest pure profit (invert sign back to positive)
            w += -heapq.heappop(max_profit_heap)

        return w
```

```python
solution = Solution()
```

```python
k = 2
w = 0
profits = [1,2,3]
capital = [0,1,1]

print(solution.findMaximizedCapital(k, w, profits, capital))
```
