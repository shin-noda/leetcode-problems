```python
class Solution(object):
    def wordPattern(self, pattern, s):
        """
        :type pattern: str
        :type s: str
        :rtype: bool
        """

        words = s.split()

        # We cannot map one-on-one (bijection)
        if len(pattern) != len(words):
            return False

        PW_map = {}
        WP_map = {}

        for p, w in zip(pattern, words):
            # Check for p to w condition
            # If p is not in PW_map, p as a value shouldn't in WP_map
            if p not in PW_map and p in WP_map.values():
                return False

            # If p is in value but w is not mapped to p
            if p in PW_map and PW_map[p] != w:
                return False

            # Check for w to p condition
            if w not in WP_map and w in PW_map.values():
                return False

            if w in WP_map and WP_map[w] != p:
                return False

            """
            print(p)
            print(w)
            print()

            print(PW_map.values())
            print(WP_map.values())
            print()
            """

            PW_map[p] = w
            WP_map[w] = p

        # print(PW_map)
        # print(WP_map)

        return True
```

```python
solution = Solution()
```
