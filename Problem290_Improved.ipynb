```python
class Solution(object):
    def wordPattern(self, pattern, s):
        """
        :type pattern: str
        :type s: str
        :rtype: bool
        """

        words = s.split()

        # We cannot map one-on-one (bijection)
        if len(pattern) != len(words):
            return False

        PW_map = {}
        WP_map = {}

        for p, w in zip(pattern, words):
            # If the mapping exists, it must match the current word
            if p in PW_map and PW_map[p] != w:
                return False

            # If the reverse mapping exists, it must match the current pattern
            if w in WP_map and WP_map[w] != p:
                return False

            # If neither exists, create the mapping
            PW_map[p] = w
            WP_map[w] = p

        return True
```

```python
solution = Solution()
```

```python
pattern = "abba"
s = "dog cat cat dog"

print(solution.wordPattern(pattern, s))
```
