```python
class Solution(object):
    def isPalindrome(self, s):
        """
        :type s: str
        :rtype: bool
        """

        n = len(s)

        # Check if s is either 0 or 1 character.
        if n <= 1:
            return True

        # Use the two pointer method
        left = 0
        right = n - 1

        while left < right:
            if s[left].isalnum() and s[right].isalnum():
                # compare
                if s[left].lower() == s[right].lower():
                    # Move both pointers
                    left = left + 1
                    right = right - 1

                else:
                    # Not a palindrome
                    return False

            else:
                if not s[left].isalnum():
                    # Move the left pointer to the right
                    left = left + 1

                if not s[right].isalnum():
                    # Move the right pointer to the left
                    right = right - 1

        return True
```

```python
solution = Solution()
```
