```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def rotateRight(self, head, k):
        """
        :type head: Optional[ListNode]
        :type k: int
        :rtype: Optional[ListNode]
        """

        # Check the edge case
        if k == 0 or not head:
            return head

        # A sentinel node
        dummy = ListNode(0, head)

        # This is the location of the last node
        last = dummy.next

        n = 0
        while head.next:
            head = head.next
            last = last.next
            n += 1

        n += 1
        last = head

        # Remainder
        # We only need to rotate r times, not k times
        r = k % n

        # To 'rotate' nodes, we need to do two things.
        # 1. Point the (n - r)th node to None, and
        # 2. Attach the last (n)th node to the first node

        # When r is 0, we don't need to do anything
        if r == 0:
            return dummy.next

        # Reset head
        head = dummy.next

        # We need to iterate n - r - 1 times
        for _ in range(n - r - 1):
            head = head.next

        # Step 1. Point the node to None.
        # We need to save the head.next node for the return value.
        result = head.next
        head.next = None

        # Step 2. Attach the last node to the first node
        last.next = dummy.next

        return result
```
