```python
class Solution(object):
    def minDistance(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: int
        """

        dp = [[float('inf')] * (len(word2) + 1) for _ in range(len(word1) + 1)]

        # Set the first element of dp as 0
        dp[0][0] = 0

        # Base case: add 1, 2, 3, ... for the first column
        for i in range(1, len(dp)):
            dp[i][0] = dp[i - 1][0] + 1

        # Base case: add 1, 2, 3, ... for the first row
        for i in range(1, len(dp[0])):
            dp[0][i] = dp[0][i - 1] + 1

        # We have 4 moves in the nested loop:
        # 1. move without paying costs (chars match) dp[i - 1][j - 1] -> dp[i][j]
        # 2. Replace: dp[i - 1][j - 1] -> dp[i][j] and you pay for 1 move
        # 3. Delete: dp[i - 1][j] -> dp[i][j] and you pay for 1 move
        # 4. Insert: dp[i][j - 1] -> dp[i][j] and you pay for 1 move

        # Fill the rest of the dp table
        for i in range(1, len(dp)):
            for j in range(1, len(dp[i])):
                # The characters of two strings match
                if word1[i - 1] == word2[j - 1]:
                    # We can move without paying the cost
                    dp[i][j] = dp[i - 1][j - 1]

                # We do either Replace, Delete, or Insert.
                else:
                    # We add 1 because it's not a cost free move
                    dp[i][j] = 1 + min(dp[i - 1][j - 1], dp[i - 1][j], dp[i][j - 1])

        return dp[-1][-1]
```

```python
solution = Solution()
```
