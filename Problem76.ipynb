```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """

        # Length of s
        m = len(s)

        # Length of t
        n = len(t)

        if n > m:
            return ""

        # Assign the 'infinite' length of string
        min_win = "" + s + t

        # print(f'min_win: {min_win}')

        # Build a hashmap (dict)
        # This is for 't'
        target_map = {}

        # This is for counting words
        current_map = {}

        for letter in t:
            target_map[letter] = target_map.get(letter, 0) + 1

        # print(f'target_map: {target_map}\n')

        # Start of window
        left = 0

        # 'right' is the end of window
        for right in range(m):
            c = s[right]

            # print(f'right: {right}')
            # print(f'c: {c}')

            if c in target_map:
                # Check the upper bound
                upper_bound = target_map.get(c)

                if c in current_map:
                    if len(current_map[c]) >= upper_bound:
                        # Pop the first value and add the new value at the end
                        current_map[c].pop(0)

                    current_map[c].append(right)

                else:
                    # Append a new value
                    current_map[c] = [right]

                # We temporarily update left
                left = right

                # Update 'left'
                # We pick the left most among the values in current_map
                for item in current_map:
                    val = current_map[item][0]

                    if val < left:
                        left = val

                # print(f'current_map: {current_map}')
                # print(f'left: {left}')
                # print(f'right: {right}')

                # Calculate the current sum
                current_sum = sum(len(v) for v in current_map.values())

                # check if current_sum has all required items
                if current_sum >= n:
                    tmp = s[left:right + 1]

                    if len(tmp) < len(min_win):
                        min_win = tmp

        # Check if the length of min_win is 'infinite'
        if len(min_win) >= (m + n):
            # We return "" when the substring is not found
            min_win = ""

        return min_win
```

```python
solution = Solution()
```
