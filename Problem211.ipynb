```python
class TrieNode(object):
    def __init__(self, children=None, is_end=False):
        self.children = [None] * 26
        self.is_end = is_end

    def _char_to_idx(self, char):
        return ord(char) - ord('a')


class WordDictionary(object):
    def __init__(self):
        self.root = TrieNode()


    def addWord(self, word):
        """
        :type word: str
        :rtype: None
        """

        head = self.root

        for c in word:
            index = head._char_to_idx(c)

            if not head.children[index]:
                node = TrieNode()
                head.children[index] = node

            # Go to the next node
            head = head.children[index]

        # This is the last word
        head.is_end = True


    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """

        head = self.root
        queue = [head]
        for c in word:
            n = len(queue)

            # Check if c is '.'
            if c == '.':
                for _ in range(n):
                    node = queue.pop(0)

                    # We check all alphabets
                    for i in range(26):
                        if node.children[i]:
                            queue.append(node.children[i])

            else:
                index = head._char_to_idx(c)

                for _ in range(n):
                    node = queue.pop(0)

                    if node.children[index]:
                        queue.append(node.children[index])

        # Check if any nodes in queue has is_end = True
        for _ in range(len(queue)):
            node = queue.pop(0)

            if node.is_end:
                return True

        return False
```
