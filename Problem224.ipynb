```python
class Solution(object):
    def calculate(self, s):
        """
        :type s: str
        :rtype: int
        """

        # Note: for this question, there is no "*" nor "/"

        # We get rid of spaces (O(N))
        s = s.replace(" ", "")

        # Only a number
        if len(s) == 1:
            return int(s)

        # LIFI (Last-In First-Out)
        # We put all numbers include positive and negative numbers and
        # simply add all of them together
        stack = []

        # This holds either 1 or -1 and a value represents the sign before
        # the bracket and will be push at '(' and be popped for ')'
        sign_stack = []

        # +1 represents '+' and -1 represents '-'
        sign = 1

        # We process the string first to put all numbers in the stack
        i = 0
        while i < len(s):
            c = s[i]

            # Retrieve number
            # Number could be multiple digits like 42, 777
            if c.isdecimal():
                # Chech the sign before the number
                tmp = 1
                if i - 1 >= 0 and s[i - 1] == '-':
                    tmp = -1

                while i + 1 < len(s) and s[i + 1].isdecimal():
                    i += 1
                    c = c + s[i]

                num = int(c)
                num = sign * tmp * num

                # Push it to the stack
                stack.append(num)

            # We change the sign
            elif c == '(':
                if i - 1 >= 0 and s[i - 1] == '-':
                    sign = sign * -1
                    sign_stack.append(-1)

                else:
                    sign_stack.append(1)

            # We change the sign
            elif c == ')':
                sign = sign * sign_stack.pop()

            i += 1

        # Adds all values in the stack
        result = 0
        for i in stack:
            result += i

        return result
```

```python
solution = Solution()
```
