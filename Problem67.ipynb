```python
class Solution(object):
    def addBinary(self, a, b):
        """
        :type a: str
        :type b: str
        :rtype: str
        """

        d_a = 0
        d_b = 0

        # Convert a into a digit number
        for i in range(len(a)):
            d_a += int(a[len(a) - 1 - i]) * pow(2, i)

        # Convert b into a digit number
        for i in range(len(b)):
            d_b += int(b[len(b) - 1 - i]) * pow(2, i)

        d_c = d_a + d_b
        result = ""

        # Convert d_c to a binary number
        while d_c != 0:
            r = int(d_c % 2)
            result = str(r) + result
            d_c = d_c // 2

        # Check if both a and b are 0's.
        if not result:
            result = "0"

        return result
```

```python
solution = Solution()
```

```python
a = "11"
b = "1"

print(solution.addBinary(a, b))
```
