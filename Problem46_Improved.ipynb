```python
class Solution(object):
    def permute(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """

        # Time Complexity: O(n * n!)
        # O(n!): n! possible mutations
        # O(n): append takes O(n)

        # Space Complexity: O(n)
        # O(n): recursion stack: the depth of our recursion tree is n

        # This performs in-place swapping

        results = []

        def backtrack(start):
            # Base case
            if start == len(nums):
                # We need slicing because we're taking a snapshot
                results.append(nums[:])
                return

            for i in range(start, len(nums)):
                # Fix the element
                # Move the elemnt at index 'i' to the 'start' position
                temp = nums[start]
                nums[start] = nums[i]
                nums[i] = temp

                # Explore
                # Recurse to fix the next position
                backtrack(start + 1)

                # Backtrack (Unfix the element)
                # Swap them back to restore the original order
                # so the 'for' loop can try the next element
                temp = nums[start]
                nums[start] = nums[i]
                nums[i] = temp

        backtrack(0)

        return results
```

```python
solution = Solution()
```

```python
nums = [1,2,3]

print(solution.permute(nums))
```

```python
nums = [0,1]

print(solution.permute(nums))
```
