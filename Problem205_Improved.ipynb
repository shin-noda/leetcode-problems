```python
class Solution(object):
    def isIsomorphic(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        # Isomorphic:
        # Two strings s and t are isomorphic
        # if the characters in s can be replaced to get t.

        # By the constraint, both s and t have the same length

        # We only need to track the mapping in both directions
        # to ensure a one-to-one mapping (bijection)

        mapST, mapTS = {}, {}

        for c1, c2 in zip(s, t):
            # If c1 was mapped to something else, or c2 was mapped to something else
            if ((c1 in mapST and mapST[c1] != c2) or
                (c2 in mapTS and mapTS[c2] != c1)):
                return False

            mapST[c1] = c2
            mapTS[c2] = c1

        return True
```

```python
solution = Solution()
```
