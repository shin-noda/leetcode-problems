```python
class Solution(object):
    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        max_val = 1
        dp = [1] * len(nums)

        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                # Strictly increasing
                if nums[i] < nums[j]:
                    dp[j] = max(dp[i] + 1, dp[j])

                    # Update the max_val
                    max_val = max(max_val, dp[j])

        return max_val
```

```python
solution = Solution()
```
