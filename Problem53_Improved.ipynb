```python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        m = float('-inf')
        tmp = 0

        for num in nums:
            tmp += num

            if tmp > m:
                m = tmp

            # If it's negative, we ignore it.
            if tmp < 0:
                tmp = 0

        return m
```

```python
solution = Solution()
```

```python
nums = [-2,1,-3,4,-1,2,1,-5,4]

print(solution.maxSubArray(nums))
```
