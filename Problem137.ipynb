```python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        ones = 0
        twos = 0

        # The first time num appears: ones becomes num, twos stays 0.
        # The sceond time num appears: ones stays 0, twos becomes 0.
        # The third time nums appears: ones stays 0, twos becomes 0.

        for num in nums:
            # Step 1: Update 'ones'
            # It gets the bit if it's in x AND NOT already in 'twos'
            # (ones XOR num) & !twos
            ones = (ones ^ num) & ~twos

            # Step 2: Update 'twos'
            # It gest the bit if it's in num AND NOT in the new 'ones
            # (twos XOR num) & !ones
            twos = (twos ^ num) & ~ones

        return ones
```

```python
solution = Solution()
```

```python
nums = [2,2,3,2]

print(solution.singleNumber(nums))
```
