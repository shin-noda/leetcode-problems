```python
class Solution(object):
    def minPathSum(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        # Number of rows (height of the grid)
        h = len(grid)

        # Number of columns (width of the grid)
        w = len(grid[0])

        # Create a 2D dp array
        dp = []
        for _ in range(h):
            tmp = []
            for _ in range(w):
                tmp.append(float('inf'))

            dp.append(tmp)

        # Initialize the first row
        dp[0][0] = grid[0][0]
        for i in range(1, w):
            dp[0][i] = dp[0][i - 1] + grid[0][i]

        # Initialize the first column
        for i in range(1, h):
            dp[i][0] = dp[i - 1][0] + grid[i][0]

        # We iterate through the subgrid
        for i in range(1, h):
            for j in range(1, w):
                # Compare the top and the left squares
                dp[i][j] = min(grid[i][j] + dp[i - 1][j], grid[i][j] + dp[i][j - 1])

        # Return the bottom right element of dp
        return dp[h - 1][w - 1]
```

```python
solution = Solution()
```
