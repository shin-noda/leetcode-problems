```python
class Solution(object):
    def insert(self, intervals, newInterval):
        """
        :type intervals: List[List[int]]
        :type newInterval: List[int]
        :rtype: List[List[int]]
        """

        # Note that you don't need to modify intervals in-place.
        # You can make a new array and return it.

        # intervals: non-overlapping and sorted in ascending order by start_i

        results = []

        # Check if intervals are empty
        if len(intervals) <= 0:
            results.append(newInterval)
            return results

        n = len(intervals)

        ns = newInterval[0]
        ne = newInterval[1]
        # Go through the intervals from the beginning
        for i in range(n):
            s = intervals[i][0]
            e = intervals[i][1]

            if ns > e:
                results.append(intervals[i])

            # Check if 'ns' is overlapping with 's' and 'e'
            elif s <= ns <= e:
                ns = min(ns, s)
                break

            # 'ns' is less than s but not overlapping
            else:
                break

        pointer = 0
        # Go through the intervals from the end
        for i in range(n - 1, -1, -1):
            s = intervals[i][0]
            e = intervals[i][1]

            # Check if 'ne' is less than s
            if ne < s:
                results.insert(len(results) + pointer, intervals[i])
                pointer -= 1

            # Check if 'ne' is overlapping with 's' and 'e'
            elif s <= ne <= e:
                ne = max(ne, e)
                break

            # 'ne' is greater than e.
            else:
                break

        # Check if it's inserted at the last or not
        if len(results) >= 1 and ns > results[-1][1]:
            results.append([ns, ne])

        else:
            results.insert(pointer, [ns, ne])

        return results
```
