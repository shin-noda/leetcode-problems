```python
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """

        # Check if grid is empty
        if not grid:
            return 0

        # Vertical length (number of rows)
        m = len(grid)

        # Horizontal length (number of columns)
        n = len(grid[0])

        coords = []
        for i in range(m):
            row = []
            for j in range(n):
                if grid[i][j] == "1":
                    row.append(-1)

                else:
                    row.append(0)

            coords.append(row)

        # This keep tracks of groups of coords
        c_map = {}
        last_key = 1

        for i in range(m):
            for j in range(n):
                # It's an island
                if coords[i][j] != 0:
                    # Default value
                    coords[i][j] = last_key

                    is_left = False
                    is_top = False

                    # Check top
                    if i - 1 >= 0 and coords[i - 1][j] != 0:
                        coords[i][j] = coords[i - 1][j]
                        is_top = True

                    # Check left
                    if j - 1 >= 0 and coords[i][j - 1] != 0:
                        coords[i][j] = coords[i][j - 1]
                        is_left = True

                        # Top and Left values are different
                        if is_top and coords[i - 1][j] != coords[i][j - 1]:
                            coords[i][j] = min(coords[i][j - 1], coords[i - 1][j])

                            prev_value = coords[i - 1][j]
                            updated_value = coords[i][j - 1]

                            # Use a smaller value as an updated key
                            if updated_value > prev_value:
                                tmp = updated_value
                                updated_value = prev_value
                                prev_value = tmp

                            # Remove coords
                            islands = c_map.pop(prev_value)

                            for island in islands:
                                x = island[0]
                                y = island[1]

                                coords[x][y] = updated_value

                            if isinstance(island, list):
                                c_map[updated_value] = c_map.get(updated_value) + islands

                            else:
                                c_map[updated_value].append(islands)

                    # Add a coordinate
                    if coords[i][j] in c_map:
                        c_map[coords[i][j]].append([i, j])

                    else:
                        # Add a new key
                        c_map[coords[i][j]] = [[i, j]]

                        last_key += 1

        return len(c_map)
```
