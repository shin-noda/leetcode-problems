```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def zigzagLevelOrder(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[List[int]]
        """

        # Check if root is empty
        if not root:
            return []

        zz_list = []

        # This is either 0 or 1 like a bit
        b = 0
        zz_queue = [root]

        while zz_queue:
            node_list = []
            n = len(zz_queue)

            for _ in range(n):
                node = zz_queue.pop(0)

                # Append from left to right
                if b == 0:
                    node_list.append(node.val)

                # Append from right to left
                else:
                    node_list.insert(0, node.val)

                if node.left:
                    zz_queue.append(node.left)

                if node.right:
                    zz_queue.append(node.right)

            zz_list.append(node_list)

            if b == 0:
                b = 1

            else:
                b = 0

        return zz_list
```
