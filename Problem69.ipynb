```python
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """

        # Note: You must not use any built-in exponent function or operator.

        sq = 0
        while sq * sq < x:
            sq += 1

        if sq * sq != x:
            return sq - 1

        return sq
```

```python
solution = Solution()
```

```python
x = 4

print(solution.mySqrt(x))
```
