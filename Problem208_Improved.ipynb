```python
class TrieNode:
    def __init__(self):
        # A fixed-size list is the most "low-level" way to do this
        self.children = [None] * 26
        self.is_end = False

class Trie(object):
    def __init__(self):
        self.root = TrieNode()


    def _char_to_idx(self, ch):
        # Constant time integer subtraction
        return ord(ch) - ord('a')


    def insert(self, word):
        """
        :type word: str
        :rtype: None
        """

        curr = self.root
        for char in word:
            idx = self._char_to_idx(char)

            if not curr.children[idx]:
                curr.children[idx] = TrieNode()

            curr = curr.children[idx]

        curr.is_end = True


    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """

        curr = self.root
        for char in word:
            idx = self._char_to_idx(char)

            if not curr.children[idx]:
                return False

            curr = curr.children[idx]

        return curr.is_end


    def startsWith(self, prefix):
        """
        :type prefix: str
        :rtype: bool
        """

        curr = self.root
        for char in prefix:
            idx = self._char_to_idx(char)

            if not curr.children[idx]:
                return False

            curr = curr.children[idx]

        return True
```
