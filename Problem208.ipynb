```python
class TreeNode(object):
    def __init__(self, val=None, char_map=None, is_end=False):
        self.val = val
        self.char_map = {}
        self.is_end = is_end


class Trie(object):
    def __init__(self):
        self.root = TreeNode()


    def insert(self, word):
        """
        :type word: str
        :rtype: None
        """

        head = self.root
        for c in word:
            # Check if c is in head's neighbors
            if c not in head.char_map:
                # Create a new TreeNode
                new_node = TreeNode(c)

                # Add a new node to a map
                head.char_map[c] = new_node

            # Go to the child node
            head = head.char_map[c]

        # This is the end of the word
        head.is_end = True


    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """

        head = self.root
        for c in word:
            if c not in head.char_map:
                return False

            head = head.char_map[c]

        return head.is_end


    def startsWith(self, prefix):
        """
        :type prefix: str
        :rtype: bool
        """

        head = self.root
        for c in prefix:
            if c not in head.char_map:
                return False

            head = head.char_map[c]

        return True
```
