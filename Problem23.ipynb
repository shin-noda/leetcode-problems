```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def mergeKLists(self, lists):
        """
        :type lists: List[Optional[ListNode]]
        :rtype: Optional[ListNode]
        """

        # Check if lists are empty
        if not lists:
            return None

        return self.mergeLists(lists)


    def mergeLists(self, lists):
        # A sentinel node
        dummy = ListNode()
        dummy.next = lists[0]

        # We use head for moving pointers
        head = dummy.next

        for i in range(1, len(lists)):
            l1 = dummy.next
            l2 = lists[i]

            # Check if l2 is empty
            if not l2:
                continue

            # Check if l1 is empty
            if not l1:
                # Update dummy.next
                dummy.next = l2
                continue

            if l1.val < l2.val:
                head = l1
                l1 = l1.next

            else:
                head = l2
                l2 = l2.next

                # Need to update dummy.next
                dummy.next = head

            while l1 and l2:
                if l1.val < l2.val:
                    head.next = l1
                    head = head.next
                    l1 = l1.next

                else:
                    head.next = l2
                    head = head.next
                    l2 = l2.next

            if l1:
                head.next = l1

            else:
                head.next = l2

        return dummy.next
```
