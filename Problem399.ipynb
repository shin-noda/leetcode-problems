```python
class WeightedNode(object):
    def __init__(self, val=None, neighbors=None, weights=None):
        self.val = val
        self.neighbors = neighbors
        self.weights = weights


class Solution(object):
    def __init__(self):
        # This holds indices of equations that the variable appears
        self.node_map = {}
        self.results = []

    def calcEquation(self, equations, values, queries):
        """
        :type equations: List[List[str]]
        :type values: List[float]
        :type queries: List[List[str]]
        :rtype: List[float]
        """

        self.initializeWeightedGraph(equations, values)
        self.solveEquations(equations, values, queries)

        return self.results


    def initializeWeightedGraph(self, equations, values):
        for i in range(len(equations)):
            v = values[i]
            equation = equations[i]
            a = equation[0]
            b = equation[1]

            node_a = None
            node_b = None

            if a in self.node_map:
                node_a = self.node_map.get(a)

            else:
                node_a = WeightedNode(a, [], [])
                self.node_map[a] = node_a

            if b in self.node_map:
                node_b = self.node_map.get(b)

            else:
                node_b = WeightedNode(b, [], [])
                self.node_map[b] = node_b

            node_a.neighbors.append(node_b)
            node_a.weights.append(v)

            node_b.neighbors.append(node_a)

            # We append the reciprocal value
            node_b.weights.append(float(1.0) / v)


    def solveEquations(self, equations, values, queries):
        for i in range(len(queries)):
            # We want to evalueate c / d
            c = queries[i][0]
            d = queries[i][1]

            result = float(-1.0)
            # If either c or d are in the map, the result is invalid.
            if c in self.node_map and d in self.node_map:
                result = self.findPath(c, d, c, 1.0, [])

                if result == 0.0:
                    # Path not found
                    result = float(-1.0)

            self.results.append(result)


    def findPath(self, start, dest, curr, w, breadcrumbs):
        # Base
        # This node was visited
        if curr in breadcrumbs:
            return float(0.0)

        # Add the path
        breadcrumbs.append(curr)

        # Path found
        if curr == dest:
            return w

        curr_node = self.node_map.get(curr)
        neighbors = curr_node.neighbors
        weights = curr_node.weights

        for i in range(len(neighbors)):
            neighbor = neighbors[i]
            weight = weights[i]

            result = self.findPath(start, dest, neighbor.val, w * weight, breadcrumbs)

            if result != 0.0:
                break

        if start in breadcrumbs and dest in breadcrumbs:
            return result

        # Couldn't find
        return float(0.0)
```

```python
equations = [["a","b"],["b","c"]]
values = [2.0,3.0]
queries = [["a","c"],["b","a"],["a","e"],["a","a"],["x","x"]]

solution = Solution()
print(solution.calcEquation(equations, values, queries))
```
