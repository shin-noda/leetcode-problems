```python
# Definition for a QuadTree node.
class Node(object):
    def __init__(self, val=False, isLeaf=False, topLeft=None, topRight=None, bottomLeft=None, bottomRight=None):
        self.val = val
        self.isLeaf = isLeaf
        self.topLeft = topLeft
        self.topRight = topRight
        self.bottomLeft = bottomLeft
        self.bottomRight = bottomRight
```

```python
class Solution(object):
    def construct(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: Node
        """

        self.grid = grid

        return self.constructQuadTree(0, 0, len(grid))


    def constructQuadTree(self, r, c, size):
        # Base case: we've reached a 1x1 square
        if size == 1:
            return Node(self.grid[r][c] == 1, True)

        # Divide: Calculate the smaller size
        half = size // 2

        # Recursively build the four quadrants
        tl = self.constructQuadTree(r, c, half)
        tr = self.constructQuadTree(r, c + half, half)
        bl = self.constructQuadTree(r + half, c, half)
        br = self.constructQuadTree(r + half, c + half, half)

        # Conquer/Combine: check if all four are leaves and have the same value
        if (tl.isLeaf and tr.isLeaf and bl.isLeaf and br.isLeaf and
            tl.val == tr.val == bl.val == br.val):

            return Node(tl.val, True)

        # If they aren't uniform, return an internal node
        # with the children attached
        return Node(True, False, tl, tr, bl, br)
```
