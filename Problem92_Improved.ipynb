```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def reverseBetween(self, head, left, right):
        """
        :type head: Optional[ListNode]
        :type left: int
        :type right: int
        :rtype: Optional[ListNode]
        """

        # We don't change anything when left = right
        if not head or left == right:
            return head

        """
        Keep one fixed pointer (the node before the reversal) and
        perform the reversal by moving nodes immediately after it.

        1. Setup: Create a dummy node pointing to head.
        2. Navigate: Move a pointer prev to the node right before the left position.
        3. Reverse (The Loop): Identify the node being moved (curr).
            - "Leapfrog" the next node over curr and place it right after prev.
            - Repeat this right - left times.
        """

        # 1. Create a sentinel node to handle the 'head' edge case
        dummy = ListNode(0)
        dummy.next = head
        prev = dummy

        # 2. Move 'prev' to the node immediately before the reversal starts
        for _ in range(left - 1):
            prev = prev.next

        # 3. Perform the reversal
        # 'curr' is the node that
        # will eventually be at the end of the reversed segment
        curr = prev.next

        for _ in range(right - left):
            # temp is the node we are currently moving to the front
            temp = curr.next

            # Cut 'temp' out by pointing 'curr' to the node after 'temp'
            curr.next = temp.next

            # Move 'temp' to the front of the sub-list (right after 'prev')
            temp.next = prev.next

            # Connect 'prev' to the new front
            prev.next = temp

        return dummy.next
```
