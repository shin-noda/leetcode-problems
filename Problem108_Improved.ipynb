```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def sortedArrayToBST(self, nums):
        """
        :type nums: List[int]
        :rtype: Optional[TreeNode]
        """

        # Height-balanced binary tree:
        # It's a binary tree in which the depth of the two subtrees of
        # every node never differs by more than one

        def createLeftAndRightNodes(left, right):
            # Base case: check if the array is empty
            if left > right:
                return None

            # Find the center
            mid = (left + right) // 2

            # Make a node
            node = TreeNode(nums[mid])

            node.left = createLeftAndRightNodes(left, mid - 1)
            node.right = createLeftAndRightNodes(mid + 1, right)

            return node

        return createLeftAndRightNodes(0, len(nums) - 1)
```

```python
solution = Solution()
```

```python
nums = [0,1,2,3,4,5,6,7]

print(solution.sortedArrayToBST(nums))
```
