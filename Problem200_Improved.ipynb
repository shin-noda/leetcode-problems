```python
class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """

        # Time Complexity: O(M * N): Visit each cell once
        # Space Complexity: O(M * N): The worst case

        # Check if grid is empty
        if not grid:
            return 0

        count = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if grid[i][j] == "1":
                    # We found a new island
                    count += 1

                    # Now sink it so we don't count its parts later
                    self.dfs(grid, i, j)

        return count

    def dfs(self, grid, r, c):
        # Boundary checks and water check
        if (
            r < 0 or
            c < 0 or
            r >= len(grid) or
            c >= len(grid[0]) or
            grid[r][c] == "0"
        ):
            return

        # Sink the current piece of land
        grid[r][c] = "0"

        # Visit all neighbours (Up, Down, Left, Right)
        self.dfs(grid, r + 1, c)
        self.dfs(grid, r - 1, c)
        self.dfs(grid, r, c + 1)
        self.dfs(grid, r, c - 1)
```

```python
solution = Solution()
```
