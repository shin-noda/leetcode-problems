```python
class Solution(object):
    def longestCommonPrefix(self, strs):
        """
        :type strs: List[str]
        :rtype: str
        """

        result = ""

        # Check if strs is empty
        if len(strs) == 0:
            return result

        # The longest common prefix won't be longer than the first word
        result = strs[0]

        # We compare from the 2nd word if it exists
        for i in range(1, len(strs)):
            word = strs[i]
            # There is no common prefix
            if len(word) == 0:
                return ""

            index = 0
            tmp = ""
            while index < len(result) and index < len(word):
                # We only need to compared with the previous result
                # because we want to find the prefix common to all words
                if result[index] != word[index]:
                    break

                else:
                    tmp = tmp + result[index]

                index = index + 1

            # Update the result
            # Use + "" to ensure that it creates a new copy of the object
            result = tmp + ""

        return result
```

```python
solution = Solution()
```

```python
strs = ["flower","flow","flight"]

print(solution.longestCommonPrefix(strs))
```

```python
strs = ["dog","racecar","car"]

print(solution.longestCommonPrefix(strs))
```
