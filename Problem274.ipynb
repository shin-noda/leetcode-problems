```python
from typing import List
```

```python
class Solution:
    def hIndex(self, citations: List[int]) -> int:
        h_index = 0

        # create an empty dictionary
        temp_dict = {}

        for citation in citations:
            # check if it's a new number
            if temp_dict.get(citation) == None:
                temp_dict[citation] = 1

            else:
                temp_dict[citation] = temp_dict[citation] + 1

        # create sorted dictionary
        sorted_list = sorted(temp_dict.items())
        dictionary = dict(sorted_list)

        sum = 0

        print(dictionary)

        print("last: " + str(int(list(dictionary.keys())[-1])))
        number = max(len(dictionary), list(dictionary.keys())[-1])

        print("number: " + str(number))

        # iterate in a reverse order
        for i in range(number, 0, -1):
            print("i: " + str(i))

            value = dictionary.get(i)

            print("value: " + str(value))

            # check if a key exists
            if value != None:
                sum = sum + value

            print("sum: " + str(sum))
            print("\n")

            if sum >= i:
                h_index = i

                return h_index

        return h_index

solution = Solution()
```
