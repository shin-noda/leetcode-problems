```python
class Solution(object):
    def findPeakElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Note: this solution's time complexity is O(N)

        # Check if there is only 1 element in nums.
        if len(nums) == 1:
            return 0

        # Check the first element
        if nums[0] > nums[1]:
            return 0

        # Check the last element
        if nums[len(nums) - 1] > nums[len(nums) - 2]:
            return len(nums) - 1

        i = 1
        n = len(nums) - 1
        while i < n:
            if nums[i] > nums[i - 1] and nums[i] > nums[i + 1]:
                return i

            if nums[i] > nums[i + 1]:
                i += 1

            i += 1
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]

print(solution.findPeakElement(nums))
```

```python
nums = [1,2,1,3,5,6,4]

print(solution.findPeakElement(nums))
```
