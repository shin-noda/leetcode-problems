```python
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        """

        # Check if head is empty or it's not connected
        if not head or not head.next:
            return False

        # Use Torsoise and Hare Algorithm
        # If there is a cycle, the faster pointer will eventually
        # lap the slow pointer and they will meet at the same node.
        # If the fast pointer hits None, you know there's no cycle

        # The slow pointer (Tortoise): Moves one step at a time
        slow = head

        # The fast pointer (Hare): Moves two steps at a time
        fast = head

        # Move fast by 2 and slow by 1
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

            # If they meet, there is a cycle
            if slow == fast:
                return True

        return False
```
