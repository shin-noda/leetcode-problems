```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: Optional[TreeNode]
        """

        # Simple definitions:
        # Preorder: root → left → right
        # Inorder: left → root → right

        # Note: both preorder and inorder are list on ints, not pointers.
        # Also, both lists consit of unique values.

        # Check if preorder is empty.
        if len(preorder) <= 0 or len(preorder) != len(inorder):
            return None

        root = TreeNode(preorder[0])

        # Create a hashmap for preorder and inorder lists
        n = len(preorder)
        pre_map = {}
        in_map = {}

        for i in range(n):
            pre_map[preorder[i]] = i
            in_map[inorder[i]] = i

        self.createBinaryTree(
            preorder,
            inorder,
            pre_map,
            in_map,
            root,
            0,
            len(preorder)
        )

        return root

    def createBinaryTree(
            self,
            preorder,
            inorder,
            pre_map,
            in_map,
            node,
            start,
            end
        ):

        pivot = in_map.get(node.val)

        # For preorder indices
        l_s = start + 1
        l_e = end

        r_s = start+ 1
        r_e = end

        is_looking = True
        # Find the left node
        for i in range(l_s, l_e):
            v = preorder[i]
            in_index = in_map.get(v)

            # We found the left node
            if is_looking and in_index < pivot:
                node.left = TreeNode(v)
                is_looking = False

            elif not is_looking and in_index > pivot:
                l_e = i
                r_s = i
                break

        is_looking = True
        # Find the right node
        for i in range(r_s, r_e):
            v = preorder[i]
            in_index = in_map.get(v)

            # We found the right node
            if is_looking and in_index > pivot:
                node.right = TreeNode(v)
                break

        if node.left:
            self.createBinaryTree(
                preorder,
                inorder,
                pre_map,
                in_map,
                node.left,
                l_s,
                l_e
            )

        if node.right:
            self.createBinaryTree(
                preorder,
                inorder,
                pre_map,
                in_map,
                node.right,
                r_s,
                r_e
            )
```
