```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """

        # Declare constants
        I = 1
        V = 5
        X = 10
        L = 50
        C = 100
        D = 500
        M = 1000

        index = 0
        roman_numerals = 0
        n = len(s)

        # There may be exceptions for I, X, and C.
        while index < n:
            char = s[index]

            match char:
                case 'I':
                    if index < n - 1:
                        next_numeral = s[index + 1]

                        if next_numeral == 'V':
                            roman_numerals = roman_numerals + (V - I)
                            index = index + 1

                        elif next_numeral == 'X':
                            roman_numerals = roman_numerals + (X - I)
                            index = index + 1

                        else:
                            roman_numerals = roman_numerals + I

                    else:
                        roman_numerals = roman_numerals + I

                case 'V':
                    roman_numerals = roman_numerals + V

                case 'X':
                    if index < n - 1:
                        next_numeral = s[index + 1]

                        if next_numeral == 'L':
                            roman_numerals = roman_numerals + (L - X)
                            index = index + 1

                        elif next_numeral == 'C':
                            roman_numerals = roman_numerals + (C - X)
                            index = index + 1

                        else:
                            roman_numerals = roman_numerals + X

                    else:
                        roman_numerals = roman_numerals + X

                case 'L':
                    roman_numerals = roman_numerals + L

                case 'C':
                    if index < n - 1:
                        next_numeral = s[index + 1]

                        if next_numeral == 'D':
                            roman_numerals = roman_numerals + (D - C)
                            index = index + 1

                        elif next_numeral == 'M':
                            roman_numerals = roman_numerals + (M - C)
                            index = index + 1

                        else:
                            roman_numerals = roman_numerals + C

                    else:
                        roman_numerals = roman_numerals + C

                case 'D':
                    roman_numerals = roman_numerals + D

                case 'M':
                    roman_numerals = roman_numerals + M

            # Increment the index
            index = index + 1

        return roman_numerals
```

```python
solution = Solution()
```

```python
s = "III"

print(solution.romanToInt(s))
```
