```python
class Solution(object):
    def summaryRanges(self, nums):
        """
        :type nums: List[int]
        :rtype: List[str]
        """

        results = []
        if len(nums) == 0:
            return results

        if len(nums) == 1:
            return [str(nums[0])]

        start = nums[0]
        interval = 0
        for i in range(1, len(nums)):
            num = nums[i]
            interval += 1

            # Not on the streak
            if num - start != interval:
                prev = nums[i - 1]

                if start == prev:
                    results.append(str(start))

                else:
                    results.append(str(start) + "->" + str(prev))

                # Update the starting point
                start = num

                # Reset the streak
                interval = 0

        # Add the last element
        if interval == 0:
            results.append(str(start))

        else:
            results.append(str(start) + "->" + str(nums[-1]))

        return results
```

```python
solution = Solution()
```

```python
nums = [0,1,2,4,5,7]

print(solution.summaryRanges(nums))
```
