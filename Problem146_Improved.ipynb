```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.val = value
        self.prev = None
        self.next = None


class LRUCache:
    def __init__(self, capacity):
        self.cap = capacity

        # Map key -> Node
        self.cache = {}

        # Sentinel nodes
        self.head = Node(0, 0)
        self.tail = Node(0, 0)
        self.head.next = self.tail
        self.tail.prev = self.head


    def _remove(self, node):
        # Unlines a node from its current position
        prev, next = node.prev, node.next
        prev.next, next.prev = prev, next

        prev.next = next
        next.prev = prev


    def _add(self, node):
        # Alwasy inserts right after the head (MRU position)
        next = self.head.next
        self.head.next = node
        node.prev = self.head
        node.next = next
        next.prev = node


    def get(self, key):
        if key in self.cache:
            node = self.cache[key]
            self._remove(node)
            self._add(node)

            return node.val

        return -1


    def put(self, key, value):
        if key in self.cache:
            self._remove(self.cache[key])

        new_node = Node(key, value)
        self.cache[key] = new_node
        self._add(new_node)

        if len(self.cache) > self.cap:
            # Evict LRU: the node right before the tail sentinel
            lru = self.tail.prev
            self._remove(lru)
            self.cache.pop(lru.key)
```
