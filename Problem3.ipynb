```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """

        n = len(s)

        if n <= 0:
            return 0

        result = 0
        max = 0
        prefix = ""
        for i in range(n):
            char = s[i]

            if s[i] not in prefix:
                # Add a char to prefix
                prefix = prefix + s[i]
                max += 1

                if max > result:
                    result = max

            # Reset the window and
            # expand the window to the left
            else:
                prefix = s[i]
                max = 1

                index = i - 1
                while index > 0:
                    if s[index] not in prefix:
                        # Add s[index] to the left of prefix
                        prefix = s[index] + prefix
                        index -= 1

                        # Increment the max
                        max += 1

                    else:
                        break

        return result
```

```python
solution = Solution()
```
