```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def isSameTree(self, p, q):
        """
        :type p: Optional[TreeNode]
        :type q: Optional[TreeNode]
        :rtype: bool
        """

        # The base case is either p or q is None.
        if not p or not q:
            # There are three cases.
            # 1. Both p and q are None (True).
            # 2. p is None and q is not None (False), and
            # 3. p is not None and q is None (False).

            return p == q

        if p.val == q.val:
            is_left_same = self.isSameTree(p.left, q.left)
            is_right_same = self.isSameTree(p.right, q.right)

        else:
            return False

        return is_left_same and is_right_same
```
