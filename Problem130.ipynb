```python
class Solution(object):
    def solve(self, board):
        """
        :type board: List[List[str]]
        :rtype: None Do not return anything, modify board in-place instead.
        """

        # Check if board is empty
        if not board:
            return

        # We check 'O's on the 4 edges

        # Check top and bottom
        for i in range(len(board[0])):
            # The 2nd param is always zero 0 because it's the top of the board
            self.dfs(board, 0, i)

            # Check if the height is greater than 1
            if len(board) > 1:
                # Bottom
                self.dfs(board, len(board) - 1, i)

        # Check left and right
        for i in range(len(board)):
            # The last param is always 0 because it's the left of the board
            self.dfs(board, i, 0)

            # Check if the width is greater than 1
            if len(board[0]) > 1:
                self.dfs(board, i, len(board[0]) - 1)

        self.processBoard(board)


    def dfs(self, board, i, j):
        # Base case:
        if (
            i < 0 or i > len(board) - 1 or
            j < 0 or j > len(board[0]) - 1 or
            board[i][j] != 'O'
        ):
            return

        # It's connected
        board[i][j] = 'C'

        # Left, right, up and down
        self.dfs(board, i - 1, j)
        self.dfs(board, i + 1, j)
        self.dfs(board, i, j - 1)
        self.dfs(board, i, j + 1)


    # This method process board so that turn all 'O' into 'X' and
    # change 'C' into 'O'.
    # We ignore 'X's.
    def processBoard(self, board):
        for i in range(len(board)):
            for j in range(len(board[0])):
                if board[i][j] == 'O':
                    board[i][j] = 'X'

                elif board[i][j] == 'C':
                    board[i][j] = 'O'
```

```python
solution = Solution()
```

```python
board = [
    ["X","X","X","X"],
    ["X","O","O","X"],
    ["X","X","O","X"],
    ["X","O","X","X"]
]

print(solution.solve(board))
```

```python
board = [
    ["X"]
]

print(solution.solve(board))
```
