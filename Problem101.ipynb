```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def isSymmetric(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: bool
        """

        # Check if root is empty
        if not root:
            return True

        # We need to start from root, not root.left and root.right.
        left_queue = [root]
        right_queue = [root]

        # We process nodes from the top to the bottom levels
        while left_queue and right_queue:
            left_vals = []
            right_vals = []

            n = len(left_queue)

            # For the left half of the tree: iterate from left to right
            # For the right half of the tree: Iterate from right to left
            for _ in range(n):
                # We process the left node first (this order doesn't matter)
                l_node = left_queue.pop(0)

                if l_node and l_node.left:
                    left_queue.append(l_node.left)
                    left_vals.append(l_node.left.val)

                else:
                    left_vals.append(None)

                if l_node and l_node.right:
                    left_queue.append(l_node.right)
                    left_vals.append(l_node.right.val)

                else:
                    left_vals.append(None)

                # Next, we process the right node
                r_node = right_queue.pop(0)

                if r_node and r_node.right:
                    right_queue.append(r_node.right)
                    right_vals.append(r_node.right.val)

                else:
                    right_vals.append(None)

                if r_node and r_node.left:
                    right_queue.append(r_node.left)
                    right_vals.append(r_node.left.val)

                else:
                    right_vals.append(None)

            if left_vals != right_vals:
                return False

        return True
```
