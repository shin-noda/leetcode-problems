```python
class Solution(object):
    def isInterleave(self, s1, s2, s3):
        """
        :type s1: str
        :type s2: str
        :type s3: str
        :rtype: bool
        """

        # Check if s3 can be made the combination of s1 and s2
        if len(s1) + len(s2) != len(s3):
            return False

        # Initialize a 2D boolean array dp of size
        dp = [[False] * (len(s2) + 1) for _ in range(len(s1) + 1)]

        # Set the first element of dp to True
        # This represents two empty strings forming an empty s3,
        # so it's always True
        dp[0][0] = True

        # Fill the first column only using s1
        for i in range(1, len(dp)):
            if dp[i - 1][0] and s1[i - 1] == s3[i - 1]:
                dp[i][0] = True

        # Fill the first row only using s2
        p2 = 0
        for i in range(1, len(dp[0])):
            if dp[0][i - 1] and s2[i - 1] == s3[i - 1]:
                dp[0][i] = True

        # Fill the rest of the table
        for i in range(1, len(dp)):
            for j in range(1, len(dp[i])):
                # Note: len(s1) + len(s2) = len(s3)
                # The incdices start from 0, so:
                # 1st character:        s3[0]
                # 2nd character:        s3[1]
                # (i + j)-th character: s3[i + j - 1]
                if (
                    dp[i - 1][j] and s1[i - 1] == s3[i + j - 1] or
                    dp[i][j - 1] and s2[j - 1] == s3[i + j - 1]
                ):
                    dp[i][j] = True

        return dp[-1][-1]
```

```python
solution = Solution()
```

```python
s1 = "aabcc"
s2 = "dbbca"
s3 = "aadbbcbcac"

print(solution.isInterleave(s1, s2, s3))
```
