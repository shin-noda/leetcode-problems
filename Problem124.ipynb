```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def maxPathSum(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        # There are two different things happening at once:
        # 1. We want the best path anywhere in the stree: stored in max_sum
        # 2. We want to return something to our parent: a single path going upward

        # Initialize global answer
        self.max_sum = float('-inf')

        # This functions answers:
        # what is the max path sum starting from this node and going up?
        def dfs(node):
            # Base case: node -> contribute nothing
            if not node:
                return 0

            # If a subtree gives a negative sum, we ignore it.
            left = max(dfs(node.left), 0)
            right = max(dfs(node.right), 0)

            # Path that splits here to update max_sum:
            # what if this node is the highest point of the path?
            current_path = left + node.val + right

            # Update global max
            self.max_sum = max(self.max_sum, current_path)

            # Return value to parent
            # Return path going up
            return node.val + max(left, right)

        dfs(root)

        return self.max_sum
```
