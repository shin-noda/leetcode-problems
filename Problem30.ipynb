```python
class Solution(object):
    def findSubstring(self, s, words):
        """
        :type s: str
        :type words: List[str]
        :rtype: List[int]
        """

        UNIT = len(words[0])
        indices = []

        # Pointer
        p = 0

        # This is the 'answer'
        # We want to match 'counter_map' to 'words_map'
        words_map = {}
        counter_map = {}
        for item in words:
            words_map[item] = words_map.get(item, 0) + 1

            # Assign 0 to all items in counter_map
            counter_map[item] = counter_map.get(item, 0)

        # print(f'words_map: {words_map}\n')

        # Special case
        if len(words_map) == 1:
            combined_string = words[0] * len(words)

            for i in range(len(s) - len(combined_string) + 1):
                if combined_string == s[i:i + len(combined_string)]:
                    indices.append(i)

        else:
            while p + UNIT * len(words) <= len(s):
                word = s[p:p + UNIT]

                # print(f'p: {p}')
                # print(f'word: {word}')

                # No match
                if word not in words_map:
                    # Reset counters
                    for item in counter_map:
                        counter_map[item] = 0

                    # Increment the pointer
                    p += 1

                # Matched
                else:
                    # Continue to check until it matches to 's' or not
                    index = p
                    is_checking = True
                    while is_checking:
                        tmp = s[index:index + UNIT]

                        # Increment the counter
                        counter_map[tmp] = counter_map.get(tmp) + 1

                        # print(f'tmp: {tmp}')
                        # print(f'counter_map: {counter_map}')

                        # Check if it's a match
                        if counter_map == words_map:
                            indices.append(p)

                            # print(f'Matched, indices: {indices}')

                            tmp = s[p + 1:p + 1 + UNIT]
                            if tmp in words_map:
                                # Shift the window by one
                                p += 1

                            else:
                                # Shift by UNIT
                                p += UNIT

                            # Reset the counters
                            for item in counter_map:
                                counter_map[item] = 0

                            # End the loop
                            is_checking = False

                        else:
                            upper_bound = words_map.get(tmp)

                            # print(f'Upper bound of {tmp}: {upper_bound}')

                            # Check if it exceeds the upper bound
                            if counter_map.get(tmp) > upper_bound:
                                # Increment the pointer by one
                                p += 1

                                # Reset counters
                                for item in counter_map:
                                    counter_map[item] = 0

                                # We end the loop
                                is_checking = False

                            else:
                                index += UNIT

                                if index + UNIT > len(s):
                                    is_checking = False

                                else:
                                    tmp = s[index:index + UNIT]

                                    # Make sure if tmp exists in the next loop
                                    if tmp  not in counter_map:
                                        p += 1

                                        # Reset counters
                                        for item in counter_map:
                                            counter_map[item] = 0

                                        is_checking = False

                # Change the line
                # print('\n')

        return indices
```
