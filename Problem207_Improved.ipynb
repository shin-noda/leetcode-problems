```python
class Solution(object):
    def canFinish(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: bool
        """

        # Use Kahn's Algorithm (BFS)

        # This algorithm detects cycles in a directed graph using topological sorting.
        # If a cycle exists, some nodes will never reach indegree 0.

        # Time Complexity: O(V + E)
        # where V is the number of courses and E is the number of prerequsites
        # Space Complexity: O(V + E)

        # 1. Calculate Indegrees: For every couse, count how many prerequisites it has.
        # 2. The "Free" Queue: Find all courses with an indegree of 0
        # (meaning they have no prerequisites) and put them in a queue.
        # 3. Process and Reduce: Pop a course from the queue, "take" it, and
        # look at all the courses that depended on it. Decrement their indegree.
        # 4. Chain Reaction: If a dependent course's indegree hits 0, add it to the queue.
        # 5. Final Check: If the number of courses you "took" equals numCourses,
        # you're good. if not, there's a circular dependency.

        # 1. Build Adjacency List and Indegree Array
        # It stores: If I finish course X, what courses unlock?
        # e.g. adj[0] = [1, 2] means course 0 unlocks 1 and 2
        adj = [[] for _ in range(numCourses)]

        # It stores: How many prerequisites does each course still need?
        # e.g. indegree[3] = 2 means course 3 needs 2 other courses first
        indegree = [0] * numCourses

        # Count how many prerequisites each course has
        for dest, src in prerequisites:
            # src must be done before dest
            adj[src].append(dest)

            # Count prerequisite
            indegree[dest] += 1

        # 2. Find starting nodes (indegree 0)
        # We collect all courses that have no prerequisites.
        queue = [i for i in range(numCourses) if indegree[i] == 0]

        # Count the number of courses we took.
        # If count == numCourses, we were able to process all courses (no cycle).
        count = 0

        # Pointer for our manual queue.
        # Avoid pop(0) (O(n)); use head pointer for O(1) queue behavior.
        head = 0

        # 3. Process the queue
        # We keep processing while there are still courses we can take
        while head < len(queue):
            curr = queue[head]
            head += 1
            count += 1

            # Unlock dependent courses
            for neighbor in adj[curr]:
                # We just finished curr, so its dependents now need 1 fewer prerequisite.
                indegree[neighbor] -= 1

                # Check if a course becomes available
                if indegree[neighbor] == 0:
                    # We can take it now
                    queue.append(neighbor)

        # 4. If we processed all nodes, no cycles exist
        return count == numCourses
```

```python
solution = Solution()
```

```python
numCourses = 2
prerequisites = [[1,0]]

print(solution.canFinish(numCourses, prerequisites))
```
