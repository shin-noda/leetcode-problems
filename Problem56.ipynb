```python
class Solution(object):
    def merge(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: List[List[int]]
        """

        """
        If I use sort(), the min Time Complexity is O(NlogN)

        But if I go through intervals (O(N)) and create a new interval and
        go through the new interval (O(N)), the complexity is O(N^2).

        Thus, I decide to sort() the intervals first.
        """

        results = []
        if len(intervals) <= 0:
            return results

        # O(NlogN)
        intervals.sort()

        results.append(intervals[0])
        for i in range(1, len(intervals)):
            start = intervals[i][0]
            end = intervals[i][1]

            # last start and end
            ls = results[-1][0]
            le = results[-1][1]

            # Case 1: No overlapping
            if start >= le + 1:
                results.append(intervals[i])

            # Case 2: Overlapping
            else:
                if le > end:
                    results[-1] = [ls, le]

                else:
                    results[-1] = [ls, end]

        return results
```

```python
solution = Solution()
```
