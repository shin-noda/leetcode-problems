```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """

        self.p = p.val
        self.q = q.val

        self.p_path = None
        self.q_path = None

        # This 1 represents the root.
        self.helper(root, "1")

        # The path to the Lowest Common Ancestor (LCA)
        lca_path = ""

        l = len(self.p_path)
        m = len(self.q_path)
        n = min(l, m)

        i = 0
        while i < n:
            if self.p_path[i] != self.q_path[i]:
                break

            else:
                lca_path = lca_path + self.p_path[i]

            i += 1

        # We ignore the root
        lca_path = lca_path[1:]

        head = root
        for str_bit in lca_path:
            bit = int(str_bit)

            if bit == 0:
                head = head.left

            else:
                head = head.right

        return head


    def helper(self, root, path):
        if not root or (self.p_path and self.q_path):
            return

        if root.val == self.p:
            self.p_path = path

        elif root.val == self.q:
            self.q_path = path

        self.helper(root.left, path + "0")
        self.helper(root.right, path + "1")
```
