```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def buildTree(self, preorder, inorder):
        """
        :type preorder: List[int]
        :type inorder: List[int]
        :rtype: Optional[TreeNode]
        """

        # Preorder: root -> left -> right
        # Inorder: left -> root -> right

        # Big Idea of the solution
        # We do two things repeatedly:
        # 1. Take next root from preorder
        # 2. Split inorder into left and right subtrees
        # Then recursively repeat.

        # Setup
        # self.pre_idx keeps track of where we are in preorder
        # We always move left -> right through preorder
        self.pre_idx = 0
        self.inorder = inorder

        # start recursion
        return self.helper(preorder, 0, len(inorder) - 1)


    # Each function call represents a complete subtree,
    # and I only need to define its root and its boudaries.
    def helper(self, preorder, left, right):
        """
        1. Pick the root from preorder
        2. Find the root in inorder
        3. Split preorder into left and right
        4. Build left subtree
        5. Buidl right subtree
        6. Return my root upward
        """

        # Base case.
        # If there are no elements left in this inorder range, return empty.
        if left > right:
            return None

        # Pick the root
        root_val = preorder[self.pre_idx]
        self.pre_idx += 1

        # Create node
        root = TreeNode(root_val)

        # find root in inorder (this is the only linear scan)
        mid = left
        while mid <= right:
            if self.inorder[mid] == root_val:
                break

            mid += 1

        root.left = self.helper(preorder, left, mid - 1)
        root.right = self.helper(preorder, mid + 1, right)

        return root
```
