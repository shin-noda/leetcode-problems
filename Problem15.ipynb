```python
class Solution(object):
    def threeSum(self, nums):
        """
        :type nums: List[int]
        :rtype: List[List[int]]
        """

        # Sort the array first
        nums.sort()

        result = []
        n = len(nums)

        # 'i' is the leftmost pointer, which is fixed in each iteration
        for i in range(n - 2):
            # Skip duplicate values for i
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left = i + 1
            right = n - 1

            # The 'i' pointer is fixed,
            # and we move the 'left' and 'right' pointers
            while left < right:
                total = nums[i] + nums[left] + nums[right]

                if total == 0:
                    result.append([nums[i], nums[left], nums[right]])

                    # Skip duplicate values for 'left'
                    while left < right and nums[left] == nums[left + 1]:
                        left += 1

                    # Skip duplicate values for 'right'
                    while left < right and nums[right] == nums[right - 1]:
                        right -= 1

                    left += 1
                    right -= 1

                elif total < 0:
                    left += 1
                else:
                    right -= 1

        return result
```

```python
solution = Solution()
```

```python
nums = [-1,0,1,2,-1,-4]

print(solution.threeSum(nums))
```

```python
nums = [0,1,1]

print(solution.threeSum(nums))
```
