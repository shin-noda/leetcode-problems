```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def sortList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # Check if head is empty
        if not head:
            return None

        # A sentinel node
        dummy = ListNode()

        add_map = {}
        num_list = []

        while head:
            num_list.append(head.val)

            if head.val not in add_map:
                add_map[head.val] = [head]

            else:
                add_map[head.val].append(head)

            # Advance the head
            head = head.next

        # Sort the list
        num_list.sort()

        # Update the sentinel node
        dummy.next = add_map[num_list[0]][0]
        head = dummy.next

        # We use a manual pointer to avoid pop(0) whose time complexity is O(n)
        pointer = 0
        temp = 0
        while pointer < len(num_list):
            val = num_list[pointer]

            node = add_map[val][temp]

            head.next = node
            head = node

            # Check if there are multiple nodes with the same value
            if pointer + 1 < len(num_list) and val == num_list[pointer + 1]:
                temp += 1

            else:
                temp = 0

            pointer += 1

        # No more link
        head.next = None

        return dummy.next
```
