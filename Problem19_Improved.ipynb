```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """

        # Create a dummy node that points to head.
        # This handles the case where we need to remove the first node.
        dummy = ListNode(0)
        dummy.next = head

        fast = dummy
        slow = dummy

        # Advance 'fast' so there are 'n' nodes between fast and slow.
        for _ in range(n):
            fast = fast.next

        # Move both until 'fast' reaches the very last node.
        # Now 'slow' is exactly behind the node we want to delete.
        while fast.next:
            fast = fast.next
            slow = slow.next

        # Skip the target node.
        slow.next = slow.next.next

        return dummy.next
```
