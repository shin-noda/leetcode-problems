```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """

        # Edge case
        if n <= 0:
            return head

        # A sentinel node
        dummy = ListNode(0)
        dummy.next = head

        # 'before' is 1 node before the target node which will be removed
        before = head

        # 'after is 1 node after the target. It could be None and that's fine.
        after = head

        # We proceed after twice
        after = after.next

        # If after is None, there is only single node
        # We proceed after later.
        if not after:
            return None

        # We gonna make the lead node (n - 1) times
        lead = head

        for _ in range(n):
            lead = lead.next

        # This is for the edge case when node is 2 and n = 2.
        if not lead:
            return after

        while lead.next:
            lead = lead.next
            before = before.next
            after = after.next

        # Now, before and after has 2 gaps and the taget node is between
        # If after is None, that's fine.
        if after:
            after = after.next

        before.next = after

        return dummy.next
```
