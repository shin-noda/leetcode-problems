```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """

        # This method ask:
        # do you have p or q somewhere under you?

        # Base case:
        # - we hit a null -> return None or
        # - find one of the nodes -> return p or q
        if not root or root == p or root == q:
            return root

        # Look for p and q in the basements
        # Did I find p or q somewhere in the left subtree?
        left = self.lowestCommonAncestor(root.left, p, q)

        # Did I find p or q somewhere in the right subtree?
        right = self.lowestCommonAncestor(root.right, p, q)

        # If both sides found something, this node is the meeting point
        if left and right:
            return root

        # Otherwise, return whichever side actually found a node
        # It's also possible that both left and right are None here.
        return left or right
```
