```python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Constraints:
        # Time Complexity: O(n)
        # Space Complexity: O(1)

        result = nums[0]
        for i in range(1, len(nums)):
            num = nums[i]
            # Use XOR operation
            result = result ^ num

        return result
```

```python
solution = Solution()
```

```python
nums = [2,2,1]

print(solution.singleNumber(nums))
```

```python
nums = [4,1,2,1,2]

print(solution.singleNumber(nums))
```
