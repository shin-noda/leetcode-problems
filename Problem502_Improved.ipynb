```python
class Solution(object):
    def findMaximizedCapital(self, k, w, profits, capital):
        """
        :type k: int
        :type w: int
        :type profits: List[int]
        :type capital: List[int]
        :rtype: int
        """

        # Prepare and sort by the capital
        # Time Complexity: O(N log N)
        n = len(profits)
        projects = []

        for i in range(n):
            # Use tuple packing
            projects.append((capital[i], profits[i]))

        # Sort by catital ascending
        projects.sort(key=lambda x:x[0])

        # Max-Heap for profitability
        # We'll use a list and iterative heapify methods
        heap = []

        def heap_push(val):
            heap.append(val)

            # Sift up
            idx = len(heap) - 1

            while idx > 0:
                parent = (idx - 1) // 2

                if heap[idx] > heap[parent]:
                    heap[idx], heap[parent] = heap[parent], heap[idx]
                    idx = parent

                else:
                    break


        def heap_pop():
            if not heap:
                return 0

            if len(heap) == 1:
                return heap.pop()

            root_val = heap[0]
            heap[0] = heap.pop()

            # Sift down
            idx = 0
            n_heap = len(heap)

            while True:
                largest = idx
                left = 2 * idx + 1
                right = 2 * idx + 2

                if left < n_heap and heap[left] > heap[largest]:
                    largest = left

                if right < n_heap and heap[right] > heap[largest]:
                    largest = right

                if largest != idx:
                    heap[idx], heap[largest] = heap[largest], heap[idx]
                    idx = largest

                else:
                    break

            return root_val


        # The Greedy Processing Loop
        # Time Compleixty: O(K log N + N log N)
        pointer = 0

        for _ in range(k):
            # Move all projects we can now afford into the Max-Heap
            while pointer < n and projects[pointer][0] <= w:
                heap_push(projects[pointer][1])
                pointer += 1

            # If the heap is empty, we can't afford any more new projects
            if not heap:
                break

            # Pick the best profit available
            w += heap_pop()

        return w
```

```python
solution = Solution()
```

```python
k = 2
w = 0
profits = [1,2,3]
capital = [0,1,1]

print(solution.findMaximizedCapital(k, w, profits, capital))
```
