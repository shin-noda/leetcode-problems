```python
class Solution(object):
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """

        is_valid = True

        # The board length
        n = 9

        # Check rows and columns
        for i in range(n):
            # We check number for each row and column,
            # so we initialize them for each 'i'
            row = set()
            column = set()

            for j in range(n):
                item_row = board[i][j]
                item_column = board[j][i]

                # Check row
                # Each row must contain the digits 1-9 without repetition
                if item_row.isnumeric():
                    if item_row in row:
                        return False

                    else:
                        row.add(item_row)

                # Check column
                # Each column must contain the digits 1-9 without repetition
                if item_column.isnumeric():
                    if item_column in column:
                        return False

                    else:
                        column.add(item_column)

                row.add(item_row)
                column.add(item_column)

        # Check for each 3 x 3 box
        # [ (0,0), (0,1), (0,2) ]
        # [ (1,0), (1,1), (1,2) ]
        # [ (2,0), (2,1), (2,2) ]
        for i in range(3):
            for j in range(3):
                box = set()

                v0 = board[i * 3][j * 3]
                v1 = board[i * 3][j * 3 + 1]
                v2 = board[i * 3][j * 3 + 2]

                v3 = board[i * 3 + 1][j * 3]
                v4 = board[i * 3 + 1][j * 3 + 1]
                v5 = board[i * 3 + 1][j * 3 + 2]

                v6 = board[i * 3 + 2][j * 3]
                v7 = board[i * 3 + 2][j * 3 + 1]
                v8 = board[i * 3 + 2][j * 3 + 2]

                if v0.isnumeric():
                    if v0 in box:
                        return False

                    else:
                        box.add(v0)

                if v1.isnumeric():
                    if v1 in box:
                        return False

                    else:
                        box.add(v1)

                if v2.isnumeric():
                    if v2 in box:
                        return False

                    else:
                        box.add(v2)

                if v3.isnumeric():
                    if v3 in box:
                        return False

                    else:
                        box.add(v3)

                if v4.isnumeric():
                    if v4 in box:
                        return False

                    else:
                        box.add(v4)

                if v5.isnumeric():
                    if v5 in box:
                        return False

                    else:
                        box.add(v5)

                if v6.isnumeric():
                    if v6 in box:
                        return False

                    else:
                        box.add(v6)

                if v7.isnumeric():
                    if v7 in box:
                        return False

                    else:
                        box.add(v7)

                if v8.isnumeric():
                    if v8 in box:
                        return False

                    else:
                        box.add(v8)

        return is_valid
```

```python
solution = Solution()
```

```python
board = [
    ["5","3",".",".","7",".",".",".","."]
    ,["6",".",".","1","9","5",".",".","."]
    ,[".","9","8",".",".",".",".","6","."]
    ,["8",".",".",".","6",".",".",".","3"]
    ,["4",".",".","8",".","3",".",".","1"]
    ,["7",".",".",".","2",".",".",".","6"]
    ,[".","6",".",".",".",".","2","8","."]
    ,[".",".",".","4","1","9",".",".","5"]
    ,[".",".",".",".","8",".",".","7","9"]
]

print(solution.isValidSudoku(board))
```
