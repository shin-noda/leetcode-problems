```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def deleteDuplicates(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # Check if it's an empty node
        if not head:
            return head

        # The sentinel node.
        dummy = ListNode(0, head)
        prev = dummy

        while head:
            # If it's a start of a duplicates range
            if head.next and head.val == head.next.val:
                # Move head until it's at the LAST node of the duplicates
                while head.next and head.val == head.next.val:
                    head = head.next

                # Skip all duplicates by linking prev to the node AFTER the last duplicate
                prev.next = head.next
            else:
                # No duplicate found for this node, so move prev forward
                prev = prev.next

            # Move head forward
            head = head.next

        return dummy.next
```
