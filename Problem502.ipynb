```python
# Profit and Capital

class PC(object):
    def __init__(self, p, c):
        self.p = p  # profit
        self.c = c  # required capital


class Solution(object):
    def findMaximizedCapital(self, k, w, profits, capital):
        """
        :type k: int
        :type w: int
        :type profits: List[int]
        :type capital: List[int]
        :rtype: int
        """

        heap = []

        def heapify_up(idx):
            parent = (idx - 1) // 2

            if idx > 0 and heap[idx].p > heap[parent].p:
                heap[idx], heap[parent] = heap[parent], heap[idx]

                heapify_up(parent)


        def max_heap(idx, heap_size):
            largest = idx

            left = 2 * idx + 1
            right = 2 * idx + 2

            if left < heap_size and heap[largest].p < heap[left].p:
                largest = left

            if right < heap_size and heap[largest].p < heap[right].p:
                largest = right

            if largest != idx:
                heap[idx], heap[largest] = heap[largest], heap[idx]

                max_heap(largest, heap_size)


        def remove_max(heap):
            if not heap:
                return None

            max_val = heap[0]

            # Move last element to root
            heap[0] = heap[-1]
            heap.pop()

            # Restore heap property
            if heap:
                max_heap(0, len(heap))

            return max_val


        # We use two sets of data:
        # 1. sorted list by capital
        # 2. max-heap based on profits -> heap[0] is the max profit

        # Create a sorted list by capital
        pcs = []
        for i in range(len(profits)):
            pcs.append(PC(profits[i], capital[i]))

        sorted_c = sorted(pcs, key=lambda x: x.c, reverse=False)

        pointer = 0
        while k > 0:
            # Check the list by capital -> create a max-heap based on profits
            for i in range(pointer, len(sorted_c)):
                if sorted_c[i].c <= w:
                    heap.append(sorted_c[i])
                    heapify_up(len(heap) - 1)

                    pointer += 1

                else:
                    break

            if heap:
                w += remove_max(heap).p

                # Decrement the number of projects we can take
                k -= 1

            else:
                # No more projects left to work on.
                break

        return w
```

```python
solution = Solution()
```

```python
k = 2
w = 0
profits = [1,2,3]
capital = [0,1,1]

print(solution.findMaximizedCapital(k, w, profits, capital))
```
