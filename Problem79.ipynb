```python
class Solution(object):
    def exist(self, board, word):
        """
        :type board: List[List[str]]
        :type word: str
        :rtype: bool
        """

        coordinates = []
        self.is_found = False

        word_set = set(word)
        array_set = set()

        for i in range(len(board)):
            for j in range(len(board[0])):
                array_set.add(board[i][j])

        if not word_set.issubset(array_set):
            return False

        def backtrack(r, c, index):
            # Base case
            if (
                not (0 <= r <= len(board) - 1) or
                not (0 <= c <= len(board[0]) - 1) or
                board[r][c] != word[index]
            ):
                return False

            # We found the word
            if (
                self.is_found or
                (board[r][c] == word[index] and index == len(word) - 1)
            ):
                self.is_found = True
                return True

            # Choose top
            if [r - 1, c] not in coordinates:
                coordinates.append([r - 1, c])

                # Move top
                backtrack(r - 1, c, index + 1)

                # Backtrack
                coordinates.pop()

            # Choose bottom
            if [r + 1, c] not in coordinates:
                coordinates.append([r + 1, c])

                # Move bottom
                backtrack(r + 1, c, index + 1)

                # Backtrack
                coordinates.pop()

            # Choose Left
            if [r, c - 1] not in coordinates:
                coordinates.append([r, c - 1])

                # Move Left
                backtrack(r, c - 1, index + 1)

                # Back track
                coordinates.pop()

            # Choose Right
            if [r, c + 1] not in coordinates:
                coordinates.append([r, c + 1])

                # Move Right
                backtrack(r, c + 1, index + 1)

                # Backtrack
                coordinates.pop()

            return self.is_found

        for i in range(len(board)):
            for j in range(len(board[0])):
                # Choose
                coordinates.append([i, j])

                # Explore
                backtrack(i, j, 0)

                # Backtrack
                coordinates.pop()

        return self.is_found
```

```python
solution = Solution()
```

```python
board = [["A","B","C","E"],["S","F","C","S"],["A","D","E","E"]]
word = "ABCCED"

print(solution.exist(board, word))
```
