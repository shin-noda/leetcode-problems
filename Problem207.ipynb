```python
class Solution(object):
    def canFinish(self, numCourses, prerequisites):
        """
        :type numCourses: int
        :type prerequisites: List[List[int]]
        :rtype: bool
        """

        # Check if the list is empty
        if not prerequisites:
            return True

        # Idea: if there is a loop created by the nodes of prerequisites,
        # it's impossible to complete courses.
        # e.g. prerequisites = [[1,0],[0,1]] -> there is a loop in nodes 0 and 1.

        # Create a graph
        graph = {}
        for (c1, c2) in prerequisites:
            if c1 not in graph:
                # Create a map within a map
                graph[c1] = {}

            # This value could be anything but avoid confusing with
            # course numbers, we use -1.
            graph[c1][c2] = -1

        # Create an inner method
        def dfs(graph, curr, dest, path):
            # Base case
            if curr in path or curr not in graph:
                return False

            path.add(curr)

            # It is reachable
            if curr == dest:
                return True

            courses = graph[curr].keys()

            for course in courses:
                is_possible = dfs(graph, course, dest, path)

                if is_possible:
                    return True

            return False

        for (c1, c2) in prerequisites:
            # To take course c1, we need to take course c2 first.
            # Therefore, if we find a path from c2 to c1, it's impossible.
            # That's why we try to find a path from c2 -> c1.
            is_path = dfs(graph, c2, c1, set())

            # There is a loop, thus we cannot complete courses.
            if is_path:
                return False

        return True
```

```python
solution = Solution()
```

```python
numCourses = 2
prerequisites = [[1,0]]

print(solution.canFinish(numCourses, prerequisites))
```

```python
numCourses = 2
prerequisites = [[1,0],[0,1]]

print(solution.canFinish(numCourses, prerequisites))
```
