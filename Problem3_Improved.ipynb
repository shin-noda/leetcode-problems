```python
class Solution(object):
    def lengthOfLongestSubstring(self, s):
        """
        :type s: str
        :rtype: int
        """

        """
        We use a Sliding Window with two pointers and
        a dictionary (hash map) to remember
        the last seen position of every character
        """

        n = len(s)

        # Stores the last seen index of each character
        char_map = {}
        max_length = 0

        # Left boundary of our window
        # We keep track of a start index.
        # If we see a character we've seen before,
        # we simply "jump" the start pointer to
        # the right of that character's last position
        start = 0
        for end in range(n):
            char = s[end]

            # If the character is in the map and within our current window
            if char in char_map and char_map[char] >= start:
                # Move the start to one position after the last occurrence
                start = char_map[char] + 1

            # Update the last seen position of the character
            char_map[char] = end

            # Calculate length of current window: (end - start + 1)
            max_length = max(max_length, end - start + 1)

        return max_length
```

```python
solution = Solution()
```
