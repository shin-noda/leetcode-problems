```python
class Solution(object):
    def canConstruct(self, ransomNote, magazine):
        """
        :type ransomNote: str
        :type magazine: str
        :rtype: bool
        """

        # We can only use each letter in 'magazine' once
        if len(magazine) < len(ransomNote):
            return False

        # Create hashmaps (dicts)
        r_map = {}
        m_map = {}

        for item in ransomNote:
            r_map[item] = r_map.get(item, 0) + 1

        for item in magazine:
            m_map[item] = m_map.get(item, 0) + 1

        # Check if ransomNote can be constructed with magazine
        for key in r_map:
            if key not in m_map:
                return False

            else:
                if m_map.get(key) < r_map.get(key):
                    return False

        return True
```

```python
solution = Solution()
```

```python
ransomNote = "a"
magazine = "b"

print(solution.canConstruct(ransomNote, magazine))
```

```python
ransomNote = "aa"
magazine = "ab"

print(solution.canConstruct(ransomNote, magazine))
```
