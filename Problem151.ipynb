```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """

        if len(s) == 0:
            return ""

        result = ""

        # Get rid of whitespaces before and after the string
        stripped_string = s.strip()

        # Split the string by space
        words = stripped_string.split()

        reversed_string = ""
        for word in words:
            reversed_string = word + " " + reversed_string

        # Get rid of the last space
        result = reversed_string.strip()

        return result
```

```python
solution = Solution()
```

```python
s = "the sky is blue"

print(solution.reverseWords(s))
```

```python
s = "  hello world  "

print(solution.reverseWords(s))
```

```python
s = "a good   example"

print(solution.reverseWords(s))
```
