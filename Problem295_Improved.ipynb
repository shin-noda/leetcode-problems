```python
# We use imports because we need to implement min-heap and max-heap,
# so the code gets really long without using the library
import heapq

class MedianFinder(object):

    def __init__(self):
        self.arr = []
        self.n = 0

        # We're gonna design this class in a way that
        # len(min_heap) >= len(max_heap) and the lengh difference
        # is at most 1.
        self.min_heap = []
        self.max_heap = []


    def addNum(self, num):
        """
        :type num: int
        :rtype: None
        """

        # Push to max_heap, then move the largest of the small half to min_heap
        heapq.heappush(self.max_heap, -num)
        heapq.heappush(self.min_heap, -heapq.heappop(self.max_heap))

        # Maintain size balance: min_heap can only be at most 1 element larger
        # than max_heap
        if len(self.min_heap) > len(self.max_heap) + 1:
            heapq.heappush(self.max_heap, -heapq.heappop(self.min_heap))


    def findMedian(self):
        """
        :rtype: float
        """

        # The length of total array is odd
        if len(self.min_heap) > len(self.max_heap):
            return float(self.min_heap[0])

        # The length of total array is even
        return (self.min_heap[0] - self.max_heap[0]) / 2.0
```
