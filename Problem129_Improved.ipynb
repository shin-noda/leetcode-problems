```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def sumNumbers(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        return self.dfs(root, 0)

    def dfs(self, node, current_sum):
        if not node:
            return 0

        # Calculate value for the current path
        current_sum = current_sum * 10 + node.val

        # If it's a leaf, return the accumulated path sum
        if not node.left and not node.right:
            return current_sum

        # Otherwise, return the sum of results from both branches
        return self.dfs(node.left, current_sum) + self.dfs(node.right, current_sum)
```
