```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def countNodes(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        num = 0
        if not root:
            return num

        # Find the leftmost node.
        num_left = 0
        head = root

        while head:
            num_left += 1
            head = head.left

        # Find the rightmost node. (This node may exist one level higher than the bottom level)
        num_right = 0
        head = root

        while head:
            num_right += 1
            head = head.right

        # A complete tree is: a perfect tree + bottom packed nodes
        for i in range(num_right):
            num += pow(2, i)

        # Check if it's a perfect tree
        if num_right == num_left:
            return num

        else:
            left = num
            right = num + pow(2, num_right)

            # Use a custom binary search to find the rightmost node exist in the bottom level
            while left <= right:
                mid = (right + left) // 2

                # The first two characters of bin() is 0b which we don't need it.
                # We also ignore the first character because which is 1 which represents the root.
                b = bin(mid)[3:]

                head = root
                for str_bit in b:
                    # It's either 0 or 1
                    bit = int(str_bit)

                    if bit == 0:
                        head = head.left

                    else:
                        head = head.right

                # Check if the node exist
                if head:
                    # Here is not mid + 1 but mid because we don't know if the next node exist.
                    num = mid
                    left = mid + 1

                else:
                    num = mid - 1
                    right = mid - 1

        return num
```
