```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """

        # Map symbols to values
        roman_map = {
            'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000
        }

        roman_numerals = 0
        n = len(s)

        for i in range(n):
            # Get the value of the current symbol
            current_value = roman_map[s[i]]

            if i + 1 < n and current_value < roman_map[s[i + 1]]:
                # If there's a next symbol and
                # it's larger than the current one,
                # we subtract the current value
                roman_numerals = roman_numerals - current_value

            else:
                roman_numerals = roman_numerals + current_value

        return roman_numerals
```

```python
solution = Solution()
```

```python
s = "III"

print(solution.romanToInt(s))
```

```python
s = "LVIII"

print(solution.romanToInt(s))
```

```python
s = "MCMXCIV"

print(solution.romanToInt(s))
```
