```python
# Definition for a Node.
class Node(object):
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
```

```python
class Solution(object):
    def cloneGraph(self, node):
        """
        :type node: Node
        :rtype: Node
        """

        # Check if node is empty
        if not node:
            return None

        # This holds the unique values of the copy of node.val
        self.node_set = set()

        # This holds addresses for nodes
        self.node_map = {}

        # A deep copy of the node
        head = Node()

        # A sentinel node
        dummy = Node()
        dummy.next = head

        self.dfs(node, head)

        return dummy.next


    def dfs(self, node, head):
        # Base case
        # This node has already been visited
        if node.val in self.node_set:
            return

        val = node.val
        head.val = val

        # Add a new unique value to the set and the hashmap
        self.node_set.add(val)
        self.node_map[val] = head

        # Update neighbor info for the head
        neighbors = node.neighbors
        head.neighbors = []
        for neighbor in neighbors:
            v = neighbor.val
            new_node = Node()
            # Check if the value is new
            if v in self.node_set:
                new_node = self.node_map.get(v)

            # Connect copied neighbors to head
            head.neighbors.append(new_node)
            self.dfs(neighbor, new_node)
```
