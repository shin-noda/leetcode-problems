```python
class Solution(object):
    def uniquePathsWithObstacles(self, obstacleGrid):
        """
        :type obstacleGrid: List[List[int]]
        :rtype: int
        """

        h = len(obstacleGrid)
        w = len(obstacleGrid[0])

        dp = []
        for _ in range(h):
            tmp = []

            for _ in range(w):
                tmp.append(0)

            dp.append(tmp)

        # Initialize the first row but when there is an obstacle, we stop.
        for i in range(w):
            if obstacleGrid[0][i] == 1:
                # There is an obstacle, so we need to stop
                break

            dp[0][i] = 1

        # Initialize the first column but when there is an obstacle, we stop.
        for i in range(h):
            if obstacleGrid[i][0] == 1:
                break

            dp[i][0] = 1

        # Go through the subgrid
        for i in range(1, h):
            for j in range(1, w):
                # We found an obstacle
                if obstacleGrid[i][j] == 1:
                    continue

                # It's the sum of the top and the left square values
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

        return dp[h - 1][w - 1]
```
