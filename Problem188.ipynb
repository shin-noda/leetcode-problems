```python
class Solution(object):
    def maxProfit(self, k, prices):
        """
        :type k: int
        :type prices: List[int]
        :rtype: int
        """

        # Check if prices is empty
        if not prices:
            return 0

        buy = [-prices[0]] * k
        sell = [0] * k

        for i in range(1, len(prices)):
            price = prices[i]

            for j in range(k):
                # For the first buy, we subtract from nothing, which is 0.
                if j == 0:
                    buy[0] = max(buy[0], 0 - price)
                    sell[0] = max(sell[0], buy[0] + price)

                # For the second buy and so on, we subtract from the previous buy.
                else:
                    buy[j] = max(buy[j], sell[j - 1] - price)
                    sell[j] = max(sell[j], buy[j] + price)

        return sell[-1]
```

```python
solution = Solution()
```
