```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """

        # Idea: since nums is originally in ascending order,
        # the peak of the value is somewhere in the middle (or edges).
        # We use binary search to find the target.

        left = 0
        right = len(nums) - 1
        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                return mid

            elif nums[mid] <= nums[right]:
                if nums[mid] < target <= nums[right]:
                    left = mid + 1

                else:
                    right = mid - 1

            # This is the case when nums[left] <= nums[mid]
            else:
                if nums[left] <= target < nums[mid]:
                    right = mid - 1

                else:
                    left = mid + 1

        return -1
```

```python
solution = Solution()
```
