```python
class Solution(object):
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """

        # Note: no '0's and '1's
        self.digit_map = {
            "2": "abc",
            "3": "def",
            "4": "ghi",
            "5": "jkl",
            "6": "mno",
            "7": "pqrs",
            "8": "tuv",
            "9": "wxyz"
        }

        self.results = ['']
        self.backtracking(digits, 0)

        return self.results


    def backtracking(self, digits, index):
        # Base case
        if index >= len(digits):
            return

        letter = self.digit_map[digits[index]]

        # This is the length of the string associated to the number.
        # e.g. 2 -> abc, so the length is 3.
        # 7 -> pqrs, so the length is 4.
        n = len(letter)

        # The unit times should be repeated to add each letter
        unit = len(self.results)

        # We create n identical copies of the items in the results
        copied_items = self.results

        for _ in range(1, n):
            self.results = self.results + copied_items

        for i in range(n):
            for j in range(unit):
                self.results[i * unit + j] = self.results[i * unit + j] + letter[i]

        # Move on to the next index of the digits
        self.backtracking(digits, index + 1)
```

```python
solution = Solution()
```

```python
digits = "23"

print(solution.letterCombinations(digits))
```

```python
digits = "2"

print(solution.letterCombinations(digits))
```

```python
digits = "234"

print(solution.letterCombinations(digits))
```
