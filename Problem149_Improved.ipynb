```python
class Solution(object):
    def maxPoints(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """

        n = len(points)
        if n <= 2:
            return n

        # We store the slop as a combination of ints,
        # so we need to know the greatest common divisor of a and b
        def get_gcd(a, b):
            while b:
                a, b = b, a % b

            return a


        max_points = 0
        for i in range(n):
            # slope_map stores {(dy, dx): count} relative to points[i]
            slope_map = {}

            # Counts the anchor point itself
            duplicate = 1

            current_max = 0

            for j in range(i + 1, n):
                dx = points[j][0] - points[i][0]
                dy = points[j][1] - points[i][1]

                # Simplify the slope using our manual GCD
                g = get_gcd(dx, dy)
                slope = (dx // g, dy // g)

                # Update the map for this specific anchor
                # Use a tuple as the key of the dict.
                slope_map[slope] = slope_map.get(slope, 0) + 1
                current_max = max(current_max, slope_map[slope])

            # The total points on a line for this anchor is
            # (points on that slope) + (the anchor point itself)
            max_points = max(max_points, current_max + duplicate)

        return max_points
```

```python
solution = Solution()
```
