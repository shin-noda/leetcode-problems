```python
class Solution(object):
    def setZeroes(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: None Do not return anything, modify matrix in-place instead.
        """

        # Vertical Length (number of rows)
        m = len(matrix)

        # Horizontal Length (number of columns)
        n = len(matrix[0])

        # We use the top row and leftmost column as markers.
        # So we use to variables to check if we need to make them 0's later
        is_top_row_zero = False
        is_left_column_zero = False

        for i in range(m):
            for j in range(n):
                cell = matrix[i][j]

                if cell == 0:
                    # Mark the topmost row above the cell as 0
                    matrix[0][j] = 0

                    # Mark the leftmost column left to the cell as 0
                    matrix[i][0] = 0

                    if i == 0:
                        is_top_row_zero = True

                    if j == 0:
                        is_left_column_zero = True

        # Overwrite the matrix value
        # Check the leftmost column but skip matrix[0][0]
        for i in range(1, m):
            cell = matrix[i][0]

            # Mark the row with 0's
            if cell == 0:
                for j in range(n):
                    matrix[i][j] = 0

        # Check the topmost row but skip matrix[0][0]
        for i in range(1, n):
            cell = matrix[0][i]

            # Mark the column with 0's
            if cell == 0:
                for j in range(m):
                    matrix[j][i] = 0

        # Overwrite the leftmost column as 0
        if is_left_column_zero:
            for i in range(m):
                matrix[i][0] = 0

        # Overwrite the topmost row as 0
        if is_top_row_zero:
            for i in range(n):
                matrix[0][i] = 0

        # print(matrix)
```

```python
solution = Solution()
```

```python
matrix = [[1,1,1],[1,0,1],[1,1,1]]

print(solution.setZeroes(matrix))
```
