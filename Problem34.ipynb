```python
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """

        # Constraint: Runtime complexity must be O(log n)
        # Use a binary search

        # Check if nums is empty
        if not nums:
            return [-1, -1]

        # First, do a binary search
        left = 0
        right = len(nums) - 1
        mid = 0

        # We find a point there it has a target value
        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                break

            elif nums[mid] < target:
                left = mid + 1

            else:
                right = mid - 1

        # Check if we found the target
        if nums[mid] != target:
            return [-1, -1]

        # We preserve those two values
        mid_backup = mid
        right_backup = right

        # Next, we check the left end of the target.
        # Note: right is always on the target value
        right = mid

        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                right = mid - 1

            elif nums[mid] < target:
                left = mid + 1

        # We append left to the result
        result = [left]

        # Finally, we find the right end of the target.
        # Note left is always on the target value

        left = mid_backup
        right = right_backup
        while left <= right:
            mid = (left + right) // 2

            if nums[mid] == target:
                left = mid + 1

            elif nums[mid] > target:
                right = mid - 1

        # We append right to the result
        result.append(right)

        return result
```
