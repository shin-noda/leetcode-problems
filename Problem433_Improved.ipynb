```python
class Solution(object):
    def minMutation(self, startGene, endGene, bank):
        """
        :type startGene: str
        :type endGene: str
        :type bank: List[str]
        :rtype: int
        """
        # Convert bank to a set for O(1) lookups
        # This is a mechanical necessity for speed
        bank_set = set(bank)

        if endGene not in bank_set:
            return -1

        # Standard list as a queue (O(n) for pop(0), but okay for small constraints)
        queue = [startGene]
        visited = {startGene}
        mutations = 0

        while queue:
            # Level-by-level traversal
            for _ in range(len(queue)):
                curr_gene = queue.pop(0)

                if curr_gene == endGene:
                    return mutations

                # Manual mutation logic
                for i in range(8):  # Length is fixed at 8
                    for char in ['A', 'C', 'G', 'T']:
                        if char == curr_gene[i]:
                            continue

                        # Reconstruct string without complex index math
                        next_gene = curr_gene[:i] + char + curr_gene[i+1:]

                        if next_gene in bank_set and next_gene not in visited:
                            visited.add(next_gene)
                            queue.append(next_gene)

            mutations += 1

        return -1
```

```python
solution = Solution()
```

```python
startGene = "AACCGGTT"
endGene = "AACCGGTA"
bank = ["AACCGGTA"]

print(solution.minMutation(startGene, endGene, bank))
```

```python
startGene = "AACCGGTT"
endGene = "AAACGGTA"
bank = ["AACCGGTA","AACCGCTA","AAACGGTA"]

print(solution.minMutation(startGene, endGene, bank))
```
