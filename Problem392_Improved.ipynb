```python
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        n_s = len(s)
        n_t = len(t)

        if n_s == 0:
            return True

        if n_s > n_t:
            return False

        # Use the two pointer method
        index_s = 0
        index_t = 0

        while index_s < n_s and index_t < n_t:
            if s[index_s] == t[index_t]:
                index_s = index_s + 1
                index_t = index_t + 1

                # Matched!
                if index_s == n_s:
                    return True

            else:
                index_t = index_t + 1

        return False
```

```python
solution = Solution()
```

```python
s = "abc"
t = "ahbgdc"

print(solution.isSubsequence(s, t))
```
