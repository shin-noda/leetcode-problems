```python
class Solution(object):
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """

        # This works but we don't do it 😉
        # result = haystack.find(needle)

        n_h, n_n = len(haystack), len(needle)

        # Edge case: empty needle -> return 0
        if n_n == 0:
            return 0

        # Needle longer than haystack -> cannot match
        if n_n > n_h:
            return -1

        # Loop over possible start indices
        for i in range(n_h - n_n + 1):
            # Check character by character
            match = True
            for j in range(n_n):
                if haystack[i + j] != needle[j]:
                    match = False
                    break

            if match:
                return i

        return -1
```

```python
solution = Solution()
```

```python
haystack = "sadbutsad"
needle = "sad"

print(solution.strStr(haystack, needle))
```
