```python
class Solution(object):
    def coinChange(self, coins, amount):
        """
        :type coins: List[int]
        :type amount: int
        :rtype: int
        """

        # Check if amount is 0
        if amount == 0:
            return 0

        dp= [float('inf')] * (amount + 1)

        # Mark the fist index as 0
        dp[0] = 0

        for i in range(amount):
            if dp[i] == float('inf'):
                continue

            for coin in coins:
                if i + coin < len(dp):
                    dp[i + coin] = min(dp[i + coin], dp[i] + 1)

        if dp[amount] == float('inf'):
            return -1

        return dp[amount]
```

```python
solution = Solution()
```

```python
coins = [1,2,5]
amount = 11

print(solution.coinChange(coins, amount))
```
