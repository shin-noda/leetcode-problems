```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Check if there is only one house.
        if len(nums) == 1:
            return nums[0]

        # check if there are two houses.
        if len(nums) == 2:
            return max(nums[0], nums[1])

        # We have at least three houses
        dp0 = nums[0]
        dp1 = max(nums[1], dp0)

        for i in range(2, len(nums)):
            if i % 2 == 0:
                dp0 = max(nums[i] + dp0, dp1)

            else:
                dp1 = max(nums[i] + dp1, dp0)

        return max(dp0, dp1)
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]

print(solution.rob(nums))
```

```python
nums = [2,7,9,3,1]

print(solution.rob(nums))
```
