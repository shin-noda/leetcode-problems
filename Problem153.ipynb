```python
class Solution(object):
    def findMin(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        left = 0
        right = len(nums) - 1

        # Check if the array is rotated.
        # If nums is rolated, nums[left] is always greater than nums[right].
        # This also covers the case when len(nums) == 1.
        if nums[left] <= nums[right]:
            return nums[left]


        while left < right:
            mid = (left + right) // 2

            if nums[left] < nums[mid]:
                if nums[right] < nums[mid]:
                    left = mid + 1

                else:
                    right = mid

            # This is the case for nums[left] >= nums[mid]
            else:
                if nums[mid] < nums[right]:
                    right = mid

                else:
                    left = mid + 1

        return nums[left]
```

```python
solution = Solution()
```
