```python
class Solution(object):
    def findMinArrowShots(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """

        """
        Implement a straightforward solution fist
        - Sort the points,
        - Merge intervals, and
        - Count the number of intervals.

        The time complexity is at least O(NlogN)
        """

        # O(NlogN)
        points.sort()

        # Check if points are empty
        if len(points) <= 0:
            return 0

        arrows = 1

        # This start and end keep tracks of the interval
        start = points[0][0]
        end = points[0][1]

        for i in range(1, len(points)):
            # The interval of a new baloon
            ps = points[i][0]
            pe = points[i][1]

            # Interval overlaps
            if (start >= ps and start <= pe) or (end >= ps and end <= pe) or (start < ps and end > pe):
                # Shrink the interval
                # Move the start to right
                start = max(start, ps)

                # Move the end to left
                end = min(end, pe)

            # Interval doesn't overlap
            else:
                # Update the new interval
                start = ps
                end = pe

                # Need a new arrow
                arrows += 1

        return arrows
```
