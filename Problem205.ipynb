```python
class Solution(object):
    def isIsomorphic(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        # Isomorphic:
        # Two strings s and t are isomorphic
        # if the characters in s can be replaced to get t.

        # By the constraint, both s and t have the same length
        n = len(s)

        s_map = {}
        t_map = {}
        s_t_map = {}

        # Map s and t simultaneously
        for i in range(n):
            item1 = s[i]
            item2 = t[i]

            # Process s
            if item1 in s_map:
                s_map[item1].append(i)

            else:
                s_map[item1] = [i]

            # Process t
            if item2 in t_map:
                t_map[item2].append(i)

            else:
                t_map[item2] = [i]

            # Create a map between s and t
            s_t_map[item1] = item2

        # Check if the key values of s and t are the same
        for key_s in s_t_map:
            key_t = s_t_map.get(key_s)

            if key_s not in s_map or key_t not in t_map:
                return False

            else:
                if s_map.get(key_s) != t_map.get(key_t):
                    return False

        return True
```

```python
solution = Solution()
```
