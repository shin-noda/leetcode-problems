```python
class Solution(object):
    def totalNQueens(self, n):
        """
        :type n: int
        :rtype: int
        """

        # Sets to track blocked paths
        cols = set()

        # We use a fact that r+c and r-c are constant for diagonal lines.
        # (r + c)
        pos_diag = set()

        # (r - c)
        neg_diag = set()

        self.count = 0

        def backtrack(r):
            # Base case: If we've placed queens in all rows
            if r == n:
                self.count += 1
                return

            for c in range(n):
                # Check if this square is under attack
                if c in cols or (r + c) in pos_diag or (r - c) in neg_diag:
                    continue

                # Place the queen
                cols.add(c)
                pos_diag.add(r + c)
                neg_diag.add(r - c)

                # Move to the next row
                backtrack(r + 1)

                # Backtrack: remove the queen to explore other options
                cols.remove(c)
                pos_diag.remove(r + c)
                neg_diag.remove(r - c)

        backtrack(0)

        return self.count
```

```python
solution = Solution()
```

```python
n = 5

print(solution.totalNQueens(n))
```
