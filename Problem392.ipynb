```python
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        n_s = len(s)
        n_t = len(t)

        if n_s == 0:
            return True

        if n_s > n_t:
            return False

        s_set = set(s)
        t_set = set(t)

        # If s is not a subset of t, it's False.
        # Both s and t are not ordered but it doesn't matter.
        if not s_set.issubset(t_set):
            return False

        index = 0
        while index + n_s <= n_t:
            tmp_s = 0
            tmp_t = 0
            # First character of s
            fs = s[tmp_s]

            if fs == t[index]:
                tmp_s = 1
                tmp_t = 1
                while tmp_s < n_s and index + tmp_t < n_t:
                    if s[tmp_s] == t[index + tmp_t]:
                        tmp_s = tmp_s + 1

                    tmp_t = tmp_t + 1

                if tmp_s >= n_s:
                    return True

            # Increment the index
            index = index + 1

        return False
```

```python
solution = Solution()
```

```python
s = "abc"
t = "ahbgdc"

print(solution.isSubsequence(s, t))
```
