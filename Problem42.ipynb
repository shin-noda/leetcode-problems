```python
class Solution(object):
    def trap(self, height):
        """
        :type height: List[int]
        :rtype: int
        """

        """
        Use the two pointer approach.
        The lower bar is the bottle neck.
        """

        total_water = 0

        # initialize two pointers
        left = 0
        right = len(height) - 1

        left_max = height[0]
        right_max = height[right]

        while left < right:
            # We always move the side that has the shorter "max" wall
            # because the shorter wall is what limits the water level.
            if left_max < right_max:
                # Move the left pointer to the right
                left = left + 1
                left_max = max(left_max, height[left])
                total_water = total_water + left_max - height[left]

            else:
                # Move the right pointer to the left
                right = right - 1
                right_max = max(right_max, height[right])
                total_water = total_water + right_max - height[right]

        return total_water
```

```python
solution = Solution()
```

```python
height = [4,2,0,3,2,5]

print(solution.trap(height))
```
