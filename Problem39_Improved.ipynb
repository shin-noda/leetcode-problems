```python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """

        # Sorting helps us "break" early if the sum exceeds target
        candidates.sort()

        results = []
        combo = []

        def backtracking(remain, start):
            # Base case
            if remain == 0:
                # The order is preserved
                results.append(list(combo))
                return

            for i in range(start, len(candidates)):
                c = candidates[i]

                # Since candidates is sorted, if c is greater than
                # the remaining target, no need to check further numbers.
                if c > remain:
                    break

                # Choose
                combo.append(c)

                # We pass 'i' as the new 'start' because
                # we can reuse the same number multiple times

                # Explore
                backtracking(remain - c, i)

                # Backtrack
                combo.pop()


        backtracking(target, 0)

        return results
```

```python
solution = Solution()
```

```python
candidates = [2,3,6,7]
target = 7

print(solution.combinationSum(candidates, target))
```
