```python
class Solution(object):
    def convert(self, s, numRows):
        """
        :type s: str
        :type numRows: int
        :rtype: str
        """

        """
        Think this problem in terms of cycles.
        And we imagine that a ball is bouncing ups and downs but
        when it bounces up, it moves to the right.
        Going down is the zig part and going up with a diagonal direction
        is the zag part.
        """

        # Check if it is empty
        if len(s) == 0:
            return ""

        n = len(s)
        # Check if it doesn't need to zig-zag
        if numRows == 1 or n <= numRows:
            return s

        result = ""
        matrix = [""] * numRows
        """
        The chars are stored in matrix in a squished way.
        It doesn't store in this way.
        P     I    N
        A   L S  I G
        Y A   H R
        P     I

        =>

        But it is stored like this.
        PIN
        ALSIG
        YAHR
        PI
        """

        targetRow = 0
        isGoingDown = True
        for i in range(n):
            # Add a single character
            matrix[targetRow] = matrix[targetRow] + s[i]

            # Decide which way it goes
            if isGoingDown:
                # This block is for going down
                if targetRow == numRows - 1:
                    # It's gonna bounce back
                    isGoingDown = False
                    targetRow = targetRow - 1

                else:
                    # Simply going down
                    targetRow = targetRow + 1

            else:
                # This block is for going up
                if targetRow == 0:
                    # It's gonna go down again
                    isGoingDown = True
                    targetRow = targetRow + 1

                else:
                    # Simply going up
                    targetRow = targetRow - 1

        result = "".join(matrix)

        return result
```

```python
solution = Solution()
```

```python
s = "PAYPALISHIRING"
numRows = 3

print(solution.convert(s, numRows))
# Correct Answer: PAHNAPLSIIGYIR
```

```python
s = "PAYPALISHIRING"
numRows = 4

print(solution.convert(s, numRows))
# Correct Answer: "PINALSIGYAHRPI"
```

```python
s = "A"
numRows = 1

print(solution.convert(s, numRows))
# Correct Answer: A
```

```python
s = "ABCDEFGH"
numRows = 2

print(solution.convert(s, numRows))
# Correct Answer: ACEGBDFH
```
