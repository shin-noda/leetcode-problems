```python
# We use imports because we need to implement min-heap and max-heap,
# so the code gets really long without using the library
import heapq

class MedianFinder(object):

    def __init__(self):
        self.arr = []
        self.n = 0

        # We're gonna design this class in a way that
        # len(min_heap) >= len(max_heap) and the lengh difference
        # is at most 1.
        self.min_heap = []
        self.max_heap = []


    def addNum(self, num):
        """
        :type num: int
        :rtype: None
        """

        if len(self.max_heap) <= len(self.min_heap):
            # Max-heap operation
            heapq.heappush(self.max_heap, -num)

            # We want to maintain that -max_heap[0] <= min_heap[0]
            while self.min_heap and -self.max_heap[0] > self.min_heap[0]:
                max_val = -heapq.heappop(self.max_heap)
                min_val = heapq.heappop(self.min_heap)

                heapq.heappush(self.max_heap, -min_val)
                heapq.heappush(self.min_heap, max_val)

        else:
            # Min-heap operation
            heapq.heappush(self.min_heap, num)

            # We want to maintain that -max_heap[0] <= min_heap[0]
            while self.max_heap and -self.max_heap[0] > self.min_heap[0]:
                max_val = -heapq.heappop(self.max_heap)
                min_val = heapq.heappop(self.min_heap)

                heapq.heappush(self.max_heap, -min_val)
                heapq.heappush(self.min_heap, max_val)

        # Increment n
        self.n += 1

    def findMedian(self):
        """
        :rtype: float
        """

        # Check if heaps are empty
        if not self.min_heap and not self.max_heap:
            return

        if self.n % 2 != 0:
            if len(self.min_heap) > len(self.max_heap):
                return self.min_heap[0]

            else:
                return -self.max_heap[0]

        return float(self.min_heap[0] - self.max_heap[0]) / 2.0
```
