```python
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """

        if x < 2:
            return x

        # We use binary search to find the number
        # Time Complexity: O(log N)

        left = 2
        right = x // 2

        while left <= right:
            # We use this formula to avoid overflow
            # (left + right) // 2 = (2left - left + right) // 2
            # = left + (right - left) // 2
            mid = left + (right - left) // 2
            num = mid * mid

            if num > x:
                right = mid - 1

            elif num < x:
                left = mid + 1

            # This is for num == x
            else:
                return mid

        # We want the floor, so right will be the largest integer
        # such that (right * right) <= x
        return right
```

```python
solution = Solution()
```
