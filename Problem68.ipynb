```python
class Solution(object):
    def fullJustify(self, words, maxWidth):
        """
        :type words: List[str]
        :type maxWidth: int
        :rtype: List[str]
        """

        # Check if words are empty
        if len(words) == 0:
            return ""

        results = []
        grouped_words = []
        line_list = []
        n = len(words)
        n_tmp = 0
        for i in range(n):
            # The length of the word
            word = words[i]
            n_w = len(word)

            # We add the space later
            if n_w + n_tmp <= maxWidth:
                line_list.append(word)

                # Increment 1 for the space
                n_tmp = n_tmp + n_w + 1

            else:
                n_tmp = 0
                copied_list = list(line_list)
                grouped_words.append(copied_list)

                line_list = []
                line_list.append(word)

                # Plus 1 for adding the space
                n_tmp = n_tmp + n_w + 1

            # Check if this is the last word
            if i >= n - 1:
                grouped_words.append(line_list)

        # We add spaces between words
        for i in range(len(grouped_words)):
            # Padded string
            ps = ""

            # Add the first word to ps
            ps = ps + grouped_words[i].pop(0)

            if len(grouped_words[i]) == 0:
                # Add spaces at the end
                ps = ps + " " * (maxWidth - len(ps))

            # Is this the last list of words?
            elif i >= len(grouped_words) - 1:
                # Add all string with " "
                ps = ps + " " + " ".join(grouped_words[i])

                # Fill the space at the end
                ps = ps + " " * (maxWidth - len(ps))

            else:
                last_word = grouped_words[i].pop()
                n_l = len(last_word)

                # Spaces between the first and the last words
                spaces = " " * (maxWidth - len(ps) - n_l)
                space_len = len(spaces)

                # This is the number of spaces between all words
                # For this variable, we don't consider how wide it has to be.
                # A_B_C <- the number is 2 in this case
                # A___B___C <- the number is also 2 as well
                n_s = len(grouped_words[i]) + 1

                # Number of space that has to be filled
                n_s_f = space_len - len("".join(grouped_words[i]))

                # n_s_f = n_s * quotient + remainder
                quotient = int(n_s_f / n_s)
                remainder = int(n_s_f % n_s)

                n_remained_words = len(grouped_words[i])
                # Add spaces
                for j in range(n_remained_words):
                    # Add spaces
                    ps = ps + " " * quotient

                    # Check if we need to add additional spaces
                    if remainder - j > 0:
                        ps = ps + " "

                    # Add the word
                    ps = ps + grouped_words[i].pop(0)

                # Add the last word
                space_for_lastword = " " * quotient
                ps = ps + space_for_lastword + last_word

            results.append(ps)

        return results
```

```python
solution = Solution()
```

```python
words = ["This", "is", "an", "example", "of", "text", "justification."]
maxWidth = 16

print(solution.fullJustify(words, maxWidth))
```
