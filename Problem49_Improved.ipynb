```python
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """

        # Time Complexity of this algorithm is O(N * K log K)
        # This stores values like
        # {'aet':['eat','tea','ate'],'ant':['tan','nat'],'abt':['bat']}
        sorted_chars_map = {}

        for s in strs:
            # Sort the string and join it back to make a string key
            # "eat" -> "aet"
            key = "".join(sorted(s))

            # If the key isn't in the dict, create an empty list first
            if key not in sorted_chars_map:
                sorted_chars_map[key] = []

            # Now we can safely append
            sorted_chars_map[key].append(s)

        # sorted_chars_map.values() returns the lists we've grouped
        return list(sorted_chars_map.values())
```

```python
solution = Solution()
```

```python
strs = ["eat","tea","tan","ate","nat","bat"]

print(solution.groupAnagrams(strs))
```

```python
strs = [""]

print(solution.groupAnagrams(strs))
```
