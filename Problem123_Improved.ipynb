```python
class Solution(object):
    def maxProfit(self, prices):
        """
        :type prices: List[int]
        :rtype: int
        """

        # Check if prices is empty
        if not prices:
            return 0

        # Initizlize states
        first_buy = -prices[0]
        first_sell = 0
        second_buy = -prices[0]
        second_sell = 0

        for i in range(1, len(prices)):
            price = prices[i]

            # Update the best time to buy the first stock
            first_buy = max(first_buy, -price)

            # Update the best profit if we sell the first stock today
            first_sell = max(first_sell, first_buy + price)

            # Update the best balance after buying the second stock
            second_buy = max(second_buy, first_sell - price)

            # Update the final profit if we sell the second stock today
            second_sell = max(second_sell, second_buy + price)

        return second_sell
```

```python
solution = Solution()
```
