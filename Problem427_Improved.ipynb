```python
# Definition for a QuadTree node.
class Node(object):
    def __init__(self, val=False, isLeaf=False, topLeft=None, topRight=None, bottomLeft=None, bottomRight=None):
        self.val = val
        self.isLeaf = isLeaf
        self.topLeft = topLeft
        self.topRight = topRight
        self.bottomLeft = bottomLeft
        self.bottomRight = bottomRight
```

```python
class Solution(object):
    def construct(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: Node
        """

        self.grid = grid

        return self.constructQuadTree(0, len(grid) - 1, 0, len(grid[0]) - 1)


    def constructQuadTree(self, t, b, l, r):
        # Base case: t == b and l == r
        # Note: since the grid is square, we only need to check t == b.
        if t == b:
            return Node((self.grid[t][r] == 1), True)

        v = (t + b) // 2
        h = (l + r) // 2
        node = Node(True, False)

        # Top Left
        node.topLeft = self.constructQuadTree(t, v, l, h)

        # Top Right
        node.topRight = self.constructQuadTree(t, v, h + 1, r)

        # Bottom Left
        node.bottomLeft = self.constructQuadTree(v + 1, b, l, h)

        # Bottom Right
        node.bottomRight = self.constructQuadTree(v + 1, b, h + 1, r)

        # Pruning
        if (node.topLeft.isLeaf and node.topRight.isLeaf and
            node.bottomLeft.isLeaf and node.bottomRight.isLeaf and
            node.topLeft.val == node.topRight.val ==
            node.bottomLeft.val == node.bottomRight.val):

            node.val = node.topLeft.val
            node.isLeaf = True

            node.topLeft = None
            node.topRight = None
            node.bottomLeft = None
            node.bottomRight = None

        return node
```
