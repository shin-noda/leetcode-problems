```python
class IslandNode(object):
    def __init__(self, val=0, left=None, right=None, top=None, bottom=None):
        self.val = val

class Solution(object):
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """

        # Check if grid is empty
        if not grid:
            return 0

        num = 0

        # Vertical length (number of rows)
        m = len(grid)

        # Horizontal length (number of columns)
        n = len(grid[0])

        last_key = 0

        # Create a grid that hold addresses of nodes
        add_grid = []
        for i in range(m):
            row = []
            for j in range(n):
                row.append([""])

            add_grid.append(row)

        # This keeps track of the group of islands
        i_map = {}

        count = 0

        # Create an address grid
        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':
                    add_grid[i][j] = IslandNode(-1)
                    count += 1

        # Connect islands
        for i in range(m):
            for j in range(n):
                if grid[i][j] == '1':

                    # Check if the island was checked before
                    if add_grid[i][j].val == -1:

                        add_grid[i][j].val = last_key
                        last_key += 1
                        i_map[add_grid[i][j].val] = [add_grid[i][j]]

                    # Check right
                    if j + 1 < n and grid[i][j + 1] == '1':
                        # Check if the right island was connected before
                        if (
                            add_grid[i][j + 1].val != -1 and
                            add_grid[i][j + 1].val != add_grid[i][j].val
                        ):
                            prev = add_grid[i][j].val
                            updated = add_grid[i][j + 1].val

                            # We want to remove the key which has a larger value
                            if updated > prev:
                                tmp = updated
                                updated = prev
                                prev = tmp

                            # Update the island group
                            self.updateIslandGroup(prev, updated, i_map)

                            # Add the bottom island
                            i_map[updated].append(add_grid[i][j + 1])

                        else:
                            # The right island is the same as the left one
                            add_grid[i][j + 1].val = add_grid[i][j].val

                            # Add the bottom island
                            i_map[add_grid[i][j].val].append(add_grid[i][j + 1])

                    #Check bottom
                    if i + 1 < m and grid[i + 1][j] == '1':
                        # The bottom island is the same as the top one
                        add_grid[i + 1][j].val = add_grid[i][j].val

                        # Add the bottom island
                        i_map[add_grid[i][j].val].append(add_grid[i + 1][j])

        return len(i_map)


    def updateIslandGroup(self, prev, updated, i_map):
        # Remove the previous island groud
        islands = i_map.pop(prev)

        for i in islands:
            # Update the island group of the previous island
            i.val = updated

        if isinstance(islands, list):
            i_map[updated] = i_map.get(updated) + islands
        else:
            i_map[updated].append(islands)
```
