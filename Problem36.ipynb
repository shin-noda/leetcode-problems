```python
class Solution(object):
    def isValidSudoku(self, board):
        """
        :type board: List[List[str]]
        :rtype: bool
        """

        def is_in_set(item, my_set):
            # Check if it's numeric
            if item.isnumeric():
                if item in my_set:
                    return False

                else:
                    my_set.add(item)

            return True

        # The board length
        n = 9

        # Check rows and columns
        for i in range(n):
            # We check number for each row and column,
            # so we initialize them for each 'i'
            row = set()
            column = set()

            for j in range(n):
                item_row = board[i][j]
                item_column = board[j][i]

                # Check row and column
                # Each row and column must contain the digits 1-9 without repetition
                if not is_in_set(item_row, row) or not is_in_set(item_column, column):
                    return False

        # Check for each 3 x 3 box
        # [ (0,0), (0,1), (0,2) ]
        # [ (1,0), (1,1), (1,2) ]
        # [ (2,0), (2,1), (2,2) ]
        for i in range(3):
            for j in range(3):
                box = set()

                box_values = [
                    board[i * 3][j * 3],
                    board[i * 3][j * 3 + 1],
                    board[i * 3][j * 3 + 2],

                    board[i * 3 + 1][j * 3],
                    board[i * 3 + 1][j * 3 + 1],
                    board[i * 3 + 1][j * 3 + 2],

                    board[i * 3 + 2][j * 3],
                    board[i * 3 + 2][j * 3 + 1],
                    board[i * 3 + 2][j * 3 + 2],
                ]

                for val in box_values:
                    if not is_in_set(val, box):
                        return False

        return True
```

```python
solution = Solution()
```

```python
board = [
    ["5","3",".",".","7",".",".",".","."]
    ,["6",".",".","1","9","5",".",".","."]
    ,[".","9","8",".",".",".",".","6","."]
    ,["8",".",".",".","6",".",".",".","3"]
    ,["4",".",".","8",".","3",".",".","1"]
    ,["7",".",".",".","2",".",".",".","6"]
    ,[".","6",".",".",".",".","2","8","."]
    ,[".",".",".","4","1","9",".",".","5"]
    ,[".",".",".",".","8",".",".","7","9"]
]

print(solution.isValidSudoku(board))
```
