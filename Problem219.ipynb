```python
class Solution(object):
    def containsNearbyDuplicate(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: bool
        """

        # We need at least 2 indices to meet the conditions
        if len(nums) < 2:
            return False

        # Since we want abs(i - j) <= k, we only want to store the last index
        num_map = {}

        for i in range(len(nums)):
            num = nums[i]

            if num in num_map:
                if abs(i - num_map.get(num)) <= k:
                    return True

                else:
                    # Store the latest index
                    num_map[num] = i

            else:
                # Use num as a key and i as a value
                num_map[num] = i

        return False
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]
k = 3

print(solution.containsNearbyDuplicate(nums, k))
```

```python
nums = [1,0,1,1]
k = 1

print(solution.containsNearbyDuplicate(nums, k))
```
