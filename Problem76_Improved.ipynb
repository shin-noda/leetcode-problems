```python
class Solution(object):
    def minWindow(self, s, t):
        # Check if either 't' or 's' are empty
        if not t or not s:
            return ""

        # Dictionary to keep a count of all the unique characters in t
        target_counts = {}
        for char in t:
            target_counts[char] = target_counts.get(char, 0) + 1

        # How many unique characters in t need to be present in the window
        required = len(target_counts)

        # window_counts stores the counts of chars in the current sliding window
        window_counts = {}

        # 'formed' tracks how many unique characters in t satisfy their required count
        formed = 0
        left = 0
        right = 0

        # Tuplet
        # min_len, start_idx, end_idx
        ans = (float("inf"), 0, 0)

        # Expand the window
        while right < len(s):
            char = s[right]
            window_counts[char] = window_counts.get(char, 0) + 1

            # If this character is part of t and its count matches the target
            if char in target_counts and window_counts[char] == target_counts[char]:
                formed += 1

            # Try to shrink the window from the left as long as it's still "valid"
            while left <= right and formed == required:
                char_left = s[left]

                # Update our best result if this window is smaller
                if (right - left + 1) < ans[0]:
                    ans = (right - left + 1, left, right)

                # Move left pointer and update window state
                window_counts[char_left] -= 1
                if char_left in target_counts and window_counts[char_left] < target_counts[char_left]:
                    formed -= 1

                left += 1

            right += 1

        if ans[0] == float("inf"):
            return ""

        return s[ans[1] : ans[2] + 1]
```

```python
solution = Solution()
```
