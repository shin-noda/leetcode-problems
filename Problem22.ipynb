```python
class Solution(object):
    def generateParenthesis(self, n):
        """
        :type n: int
        :rtype: List[str]
        """

        results = {'()'}

        def insertParenthesis(n):
            # Base case
            if n == 1:
                return

            parentheses = list(results)

            # clear the results
            results.clear()

            for par in parentheses:
                # This num is the number of indexes that () is inserted
                num = len(par) + 1

                for i in range(num):
                    results.add(par[:i] + '()' + par[i:])

            # Recurse
            insertParenthesis(n - 1)

        insertParenthesis(n)

        return list(results)
```

```python
solution = Solution()
```

```python
n = 3

print(solution.generateParenthesis(n))
```

```python
n = 1

print(solution.generateParenthesis(n))
```
