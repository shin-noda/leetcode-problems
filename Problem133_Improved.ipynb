```python
# Definition for a Node.
class Node(object):
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
```

```python
class Solution(object):
    def __init__(self):
        # Maps original node.val -> cloned Node instance
        self.visited = {}

    def cloneGraph(self, node):
        """
        :type node: Node
        :rtype: Node
        """

        # Check if node is empty
        if not node:
            return None

        # If we already cloned this node, just return the reference
        if node.val in self.visited:
            return self.visited[node.val]

        # Create the clone (without neighbors yet)
        clone = Node(node.val, [])

        # Store it in the map before visiting neighbors
        self.visited[node.val] = clone

        # Recursively clone and add neighbors
        for neighbor in node.neighbors:
            # We append the result of the DFS call to our list
            clone.neighbors.append(self.cloneGraph(neighbor))

        return clone
```
