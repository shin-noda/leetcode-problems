```python
class Solution(object):
    def trailingZeroes(self, n):
        """
        :type n: int
        :rtype: int
        """

        # Idea: to have trailing 0, we need at least one 2 and one 5.
        # Similarly, if we need two trailing zeros, we need at least two 2's and 5's.
        # However, number of even numbers is always is greater than the
        # number of five's so we just count the number of 5's.

        counter = 0
        num = 5

        while num <= n:
            counter = counter + (n // num)
            num *= 5

        return counter
```

```python
solution = Solution()
```

```python
n = 3

print(solution.trailingZeroes(n))
```
