```python
class Solution(object):
    def isValid(self, s):
        """
        :type s: str
        :rtype: bool
        """

        # Note: s consists of parentheses only '()[]{}'.

        # LIFO (Last in First Out)
        stack = []

        for char in s:
            if char == '(' or char == '{' or char == '[':
                stack.append(char)

            else:
                # Check if it's empty
                if len(stack) <= 0:
                    return False

                # Pop the last item
                item = stack.pop()

                if (item == '(' and char != ')') or (item == '{' and char != '}') or (item == '[' and char != ']'):
                    return False

        # It's a valid parentheses since all of them were paired.
        if len(stack) <= 0:
            return True

        return False
```

```python
solution = Solution()
```

```python
s = "()"

print(solution.isValid(s))
```

```python
s = "()[]{}"

print(solution.isValid(s))
```

```python
s = "(]"

print(solution.isValid(s))
```

```python
s = "([])"

print(solution.isValid(s))
```

```python
s = "([)]"

print(solution.isValid(s))
```
