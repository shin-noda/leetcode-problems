```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def mergeTwoLists(self, list1, list2):
        """
        :type list1: Optional[ListNode]
        :type list2: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # The dummy node keeps track of the start
        dummy = ListNode()
        current = dummy

        # While both lists have nodes, compare and pick the smaller one
        while list1 and list2:
            if list1.val <= list2.val:
                # Splicing: point to the existing node
                current.next = list1

                # Move list2 forward
                list1 = list1.next

            else:
                # Splicing: point to the existing node
                current.next = list2

                # Move list2 forward
                list2 = list2.next

            # Move our "builder" pointer forward
            current = current.next

        # The "Leftover" Trick: this returs the first True value
        # If one list is finished, just attach the rest of the other list.
        # This is way faster than looping through the remainder.
        current.next = list1 or list2

        return dummy.next
```
