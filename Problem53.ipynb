```python
class Solution(object):
    def maxSubArray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        s = sum(nums)

        # This is for determining the last index of the substring
        # It checks from left to right
        ml = float('-inf')
        tmpl = 0
        li = 0


        # This is for determining the first index of the substring
        # It checks from right to left
        mf = float('-inf')
        tmpf = 0
        fi = 0

        # We check from left to right and
        # from right to left
        for i in range(len(nums)):
            tmpl = tmpl + nums[i]
            tmpf = tmpf + nums[len(nums) - 1 - i]

            # Note: either
            # - tmpl >= ml and tmpf > mf or
            # - tmpl > ml and tmpf >= mf
            # should be choses. One of the conditions have to be >=.
            if tmpl >= ml:
                ml = tmpl
                li = i

            if tmpf > mf:
                mf = tmpf
                fi = len(nums) - 1 - i

            # If they are negative, assign 0.
            tmpl = max(0, tmpl)
            tmpf = max(0, tmpf)

        return sum(nums[fi:li + 1])
```

```python
solution = Solution()
```
