```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.word = None


class Solution(object):
    def findWords(self, board, words):
        """
        :type board: List[List[str]]
        :type words: List[str]
        :rtype: List[str]
        """

        # Build the Trie
        root = TrieNode()
        for w in words:
            node = root

            for char in w:
                if char not in node.children:
                    node.children[char] = TrieNode()

                node = node.children[char]

            node.word = w

        self.results = []
        self.rows = len(board)
        self.cols = len(board[0])

        # Start DFS from every cell
        for r in range(self.rows):
            for c in range(self.cols):
                if board[r][c] in root.children:
                    self.backtrack(r, c, root, board)

        return self.results


    def backtrack(self, r, c, parent, board):
        char = board[r][c]

        # Proceed to a child node.
        curr_node = parent.children[char]

        # If we found a word, add it and clear it to avoid duplicates
        if curr_node.word:
            self.results.append(curr_node.word)

            # Remove the word from the node so we don't pick it up twice
            curr_node.word = None

        # Mark the cell as visited
        board[r][c] = "#"

        # Explore neighbors: Right (0, 1), Left (0, -1), Up (1, 0), Down (-1, 0)
        dr = [0, 0, 1, -1]
        dc = [1, -1, 0, 0]

        for i in range(4):
            nr = r + dr[i]
            nc = c + dc[i]

            # Boundary Check
            if 0 <= nr < self.rows and 0 <= nc < self.cols:
                next_char = board[nr][nc]

                # Trie Check (Is this path worth walking?)
                if next_char in curr_node.children:
                    # Move to the child node in the Trie and recurse
                    self.backtrack(nr, nc, curr_node, board)

        # Restoration: the 'backtrack' step (Reset the board for the next path)
        board[r][c] = char

        # Pruning: if this node is a dead end, remove it
        # If curr_node.children is None, it means
        # there's no paths connected to curr_node.
        if not curr_node.children:
            parent.children.pop(char)
```

```python
solution = Solution()
```

```python
board = [
    ["o","a","a","n"],
    ["e","t","a","e"],
    ["i","h","k","r"],
    ["i","f","l","v"]
]
words = ["oath","pea","eat","rain"]

print(solution.findWords(board, words))
```
