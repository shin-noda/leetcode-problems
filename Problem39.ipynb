```python
class Solution(object):
    def combinationSum(self, candidates, target):
        """
        :type candidates: List[int]
        :type target: int
        :rtype: List[List[int]]
        """

        # Sort candinates from a small to a large number
        # O(Nlog(N))
        candidates.sort()

        # No possible combinations
        if not candidates or candidates[0] > target:
            return []

        results = []
        combo = []

        def backtracking(sum):
            # Base case
            if sum >= target:
                if sum == target:
                    # We need to sort it.
                    copied = combo[:]
                    copied.sort()

                    if copied not in results:
                        # We need a slicing to avoid reference problems
                        results.append(copied)

                return

            for c in candidates:
                # Append
                combo.append(c)
                sum += c

                # Explore
                backtracking(sum)

                # Backtrack
                combo.pop()
                sum -= c

        backtracking(0)

        return results
```

```python
solution = Solution()
```

```python
candidates = [2,3,6,7]
target = 7

print(solution.combinationSum(candidates, target))
```
