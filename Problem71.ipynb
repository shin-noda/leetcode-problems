```python
class Solution(object):
    def simplifyPath(self, path):
        """
        :type path: str
        :rtype: str
        """

        # LIFO (Last-In First-Out)
        stack = []

        dirs = path.split('/')

        for i in range(1, len(dirs)):
            item = dirs[i]

            # If 'item' is '' or '.', we ignore them.
            if item != '' and item != '.':
                # We go to a parent directory
                if item == '..':
                    # We cannot go above the root directory
                    if len(stack) != 0:
                        # We go to the parent directory
                        stack.pop()

                # Regular direcotry
                else:
                    # Add a directory
                    stack.append(item)

        # We concatinate items with '/'
        # Add a root directory with "/" +
        results = "/" + "/".join(stack)

        return results
```

```python
solution = Solution()
```

```python
path = "/home/"

print(solution.simplifyPath(path))
```
