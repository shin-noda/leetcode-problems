```python
class Solution(object):
    def minimumTotal(self, triangle):
        """
        :type triangle: List[List[int]]
        :rtype: int
        """

        dp = [float('inf')] * len(triangle[-1])

        # We mark the first element of dp with the first element of triangle
        dp[0] = triangle[0][0]

        for i in range(len(triangle) - 1):
            # This loops backwards to avoid overwriting mistakes
            for j in range(len(triangle[i]) - 1, -1, -1):
                # No overlapping from the top nodes
                if j == len(triangle[i]) - 1:
                    # We update backwards: Update dp[j + 1] first then dp[j]
                    dp[j + 1] = dp[j] + triangle[i + 1][j + 1]
                    dp[j] = dp[j] + triangle[i + 1][j]

                else:
                    # The bottom one is overlapping with the previous path
                    # We compare with the previous element
                    dp[j + 1] = min(
                        dp[j] + triangle[i + 1][j + 1],
                        dp[j + 1]
                    )
                    dp[j] = dp[j] + triangle[i + 1][j]

        return min(dp)
```

```python
solution = Solution()
```

```python
triangle = [[2],[3,4],[6,5,7],[4,1,8,3]]

print(solution.minimumTotal(triangle))
```

```python
triangle = [[-10]]

print(solution.minimumTotal(triangle))
```
