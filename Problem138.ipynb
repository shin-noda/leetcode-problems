```python
# Definition for a Node.
class Node:
    def __init__(self, x, next=None, random=None):
        self.val = int(x)
        self.next = next
        self.random = random
```

```python
class Solution(object):
    def copyRandomList(self, head):
        """
        :type head: Node
        :rtype: Node
        """

        dummy = Node(0, None, None)
        current = dummy

        # Make a copy of the starting point
        start = head

        # This map stores pairs of mirroring nodes
        address_map = {}

        while head:
            current.next = Node(head.val)

            # Move forward
            current = current.next

            address_map[head] = current

            # Move forward
            head = head.next

        # We go through the loop again
        while start:
            current = address_map.get(start)

            if start.random:
                current.random = address_map.get(start.random)

            start = start.next

        return dummy.next
```
