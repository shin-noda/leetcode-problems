```python
class Solution(object):
    def lengthOfLIS(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Check if nums is empty
        if not nums:
            return 0

        # tails[i] is the smallest tail of
        # all increasing subsequences of lenght i + 1
        # Note: tails array is NOT a squence tracker but a length tracker.
        tails = []

        for num in nums:
            # Check if I can make my current longest sequence even longer
            # Since tails is always kept in sorted order,
            # tails[-1] is the largest value we've found so far that ends a valid sequence.
            if not tails or num > tails[-1]:
                tails.append(num)

            else:
                # Manual Binary Search to find the insertion point
                low = 0
                high = len(tails) - 1
                while low <= high:
                    mid = (low + high) // 2

                    if tails[mid] < num:
                        low = mid + 1

                    else:
                        high = mid - 1

                # low is the index of the first element >= num
                tails[low] = num

        return len(tails)
```

```python
solution = Solution()
```
