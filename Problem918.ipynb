```python
class Solution(object):
    def maxSubarraySumCircular(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Max
        ma = float('-inf')

        # Min
        mi = float('inf')

        # This is the sum
        su = 0

        tp_ma = 0
        tp_mi = 0
        for num in nums:
            su += num

            tp_ma += num
            tp_mi += num

            if tp_ma > ma:
                ma = tp_ma

            if tp_mi < mi:
                mi = tp_mi

            if tp_ma < 0:
                tp_ma = 0

            if tp_mi > 0:
                tp_mi = 0

        result = ma

        if su - mi > 0:
            result = max(ma, su - mi)

        # Whichever larger is the max
        return result
```

```python
solution = Solution()
```

```python
nums = [1,-2,3,-2]

print(solution.maxSubarraySumCircular(nums))
```
