```python
class Solution(object):
    def candy(self, ratings):
        """
        :type ratings: List[int]
        :rtype: int
        """
        number_of_candies = len(ratings)

        # initialize the array
        # everyone gets at least one candy
        candies = [1] * len(ratings)

        # check if there is only one child
        if number_of_candies == 1:
             return number_of_candies

        # check negithbours from left to right and right to left.
        # when it compares from left to right, it should add the candy to the right one
        # and don't change the left one and vice versa for the oposite direction.

        # check from left to right
        for x in range(len(ratings) - 1):
            if ratings[x] < ratings[x + 1]:
                candies[x + 1] = candies[x] + 1

        # check from right to left
        for x in range(len(ratings) - 1):
            if ratings[len(ratings) - x - 2] > ratings[len(ratings) - x - 1] and candies[len(ratings) - x - 1] >= candies[len(ratings) - x - 2]:
                candies[len(ratings) - x - 2] = candies[len(ratings) - x - 1] + 1

        # reset number_of_candies
        number_of_candies = 0

        for x in candies:
            number_of_candies = number_of_candies + x

        return number_of_candies
```

```python
solution = Solution()
```

```python
ratings = [1,0,2]

print(solution.candy(ratings))
```
