```python
class Solution(object):
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """

        n = len(s)
        if n <= 1:
            return s

        # Initialize a table of False
        dp = [[False] * n for _ in range(n)]
        longest_sub = s[0]

        # Base case: length 1
        for i in range(n):
            # A diagonal line
            dp[i][i] = True

        # Base case: length 2
        for i in range(n - 1):
            if s[i] == s[i + 1]:
                dp[i][i + 1] = True
                longest_sub = s[i:i + 2]

        # The chain reaction: length 3 to n
        # We start at length 3 because we already solved 1 and 2
        for length in range(3, n + 1):
            for i in range(n - length + 1):
                # End index
                j = i + length - 1

                # Ends match and the inner core is a palindrome
                if s[i] == s[j] and dp[i + 1][j - 1]:
                    dp[i][j] = True
                    longest_sub = s[i:j + 1]

        return longest_sub
```
