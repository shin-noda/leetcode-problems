```python
class Solution(object):
    def kSmallestPairs(self, nums1, nums2, k):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :type k: int
        :rtype: List[List[int]]
        """

        heap = []
        for i in range(k):
            heap.append((float('inf'), float('inf')))


        def heapify_up(idx):
            parent = (idx - 1) // 2

            # Max heap: parent value should be greater than its child value
            if idx > 0 and sum(heap[idx]) > sum(heap[parent]):
                heap[idx], heap[parent] = heap[parent], heap[idx]

                # Sift up
                heapify_up(parent)


        def heap_push(t):
            heap.append(t)
            heapify_up(len(heap) - 1)


        def heapify_down(idx, heap_size):
            largest = idx

            left = 2 * idx + 1
            right = 2 * idx + 2

            # Max-Heap: parent >= child
            if left < heap_size and sum(heap[left]) > sum(heap[largest]):
                largest = left

            if right < heap_size and sum(heap[right]) > sum(heap[largest]):
                largest = right

            if idx != largest:
                heap[idx], heap[largest] = heap[largest], heap[idx]

                # Sift down
                heapify_down(largest, heap_size)


        def remove_heap():
            if not heap:
                return None

            if len(heap) == 1:
                return heap.pop()

            root = heap[0]

            # Put the past value of heap at the top
            heap[0] = heap.pop()

            # Sift down
            heapify_down(0, len(heap))

            return root


        # Use double loop to cover the all combinations of nums1 and nums2
        for i in range(len(nums1)):
            if (nums1[i] + nums2[0]) >= sum(heap[0]):
                break

            for j in range(len(nums2)):
                if sum(heap[0]) > (nums1[i] + nums2[j]):
                    # We add the value to the heap
                    heap_push((nums1[i], nums2[j]))

                    # Remove the first element
                    remove_heap()

                else:
                    continue

        # heap is a list of tuples so we're gonna convert it to a list of lists.
        return [list(t) for t in heap]
```

```python
solution = Solution()
```

```python
nums1 = [1,7,11]
nums2 = [2,4,6]
k = 3

print(solution.kSmallestPairs(nums1, nums2, k))
```
