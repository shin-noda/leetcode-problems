```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        # This approaches is a bottom-up approach

        # Base Case: If the node is empty, depth is 0
        if not root:
            return 0

        # Recursive Step: Get the depth of both subtrees
        # Ask the left child its depth
        left_depth = self.maxDepth(root.left)

        # Ask the right child its depth
        right_depth = self.maxDepth(root.right)

        # Manual max comparison
        if left_depth > right_depth:
            # +1 to count the current node
            return left_depth + 1

        else:
            # +1 to count the current node
            return right_depth + 1
```
