```python
class Solution(object):
    def maxPoints(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """

        # There is only one point
        if len(points) == 1:
            return 1

        slope_map = {}
        # The slope of the line: dy / dx
        m = 0

        # The y intersect of the line
        b = 0
        for i in range(len(points)):
            for j in range(i + 1, len(points)):
                # the slope m is dy / dx

                # Check if the slope is flat
                if points[j][1] == points[i][1]:
                    m = 0
                    b = points[i][1]

                # Check if the slop is a vertical line
                elif points[j][0] == points[i][0]:
                    m = float('inf')
                    b = points[i][0]

                else:
                    m = float(points[j][1] - points[i][1]) / (points[j][0] - points[i][0])
                    b = points[i][1] - m * points[i][0]

                if m not in slope_map:
                    slope_map[m] = {}
                    slope_map[m][b] = [points[i], points[j]]

                elif b not in slope_map[m]:
                    slope_map[m][b] = [points[i], points[j]]

                else:
                    if points[i] not in slope_map[m][b]:
                        slope_map[m][b].append(points[i])

                    if points[j] not in slope_map[m][b]:
                        slope_map[m][b].append(points[j])

        result = 0
        # Iterate through the nested map
        for inner_map in slope_map.values():
            for value in inner_map.values():
                if result < len(value):
                    result = len(value)

        return result
```
