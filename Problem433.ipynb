```python
class Solution(object):
    def minMutation(self, startGene, endGene, bank):
        """
        :type startGene: str
        :type endGene: str
        :type bank: List[str]
        :rtype: int
        """

        # Check if bank is empty or endGene is not in bank
        if not bank or endGene not in bank:
            return -1

        mutation = 0
        dna = ['A', 'C', 'G', 'T']
        dna_map = {'A': 0, 'C': 1, 'G': 2, 'T': 3}
        queue = [startGene]

        # This is for avoid loops
        gene_set = set([startGene])

        while queue:
            n = len(queue)
            mutation += 1

            for i in range(n):
                gene = queue.pop(0)
                for j in range(len(startGene)):
                    for k in range(len(dna) - 1):
                        # We need to mutate the gene.
                        # First, we get the index of the gene from dna
                        index = dna_map.get(gene[j])

                        # Next, we add k + 1 to the index. It ranges from 1 to 3
                        # e.g. index for G: index = 2
                        # Mutate once -> index = 3. So it's T
                        # Mutate twice -> index = 4->0. So it's A
                        # Mutate three times: index=5->1. So it's C
                        index += k + 1
                        index = index % 4

                        mutated_gene = dna[index]

                        new_gene = mutated_gene

                        if j > 0:
                            new_gene = gene[:j] + new_gene

                        if j + 1 < len(startGene):
                            new_gene = new_gene + gene[j + 1:]

                        if new_gene in bank and new_gene not in gene_set:
                            gene_set.add(new_gene)
                            queue.append(new_gene)

                        if new_gene == endGene:
                            return mutation

        return -1
```

```python
solution = Solution()
```

```python
startGene = "AACCGGTT"
endGene = "AACCGGTA"
bank = ["AACCGGTA"]

print(solution.minMutation(startGene, endGene, bank))
```
