```python
class Solution(object):
    def evalRPN(self, tokens):
        """
        :type tokens: List[str]
        :rtype: int
        """

        # LIFO (Last-In First-Out)
        stack = []

        for t in tokens:
            if t == '+':
                n2 = stack.pop()
                n1 = stack.pop()

                stack.append(n1 + n2)

            elif t == '-':
                n2 = stack.pop()
                n1 = stack.pop()

                stack.append(n1 - n2)

            elif t == '*':
                n2 = stack.pop()
                n1 = stack.pop()

                stack.append(n1 * n2)

            elif t == '/':
                n2 = stack.pop()
                n1 = stack.pop()

                # We need this for Python 2
                stack.append(int(float(n1) / n2))

            else:
                stack.append(int(t))

        # There will be only one element in the stack at the end
        result = stack.pop()

        return result
```

```python
solution = Solution()
```

```python
tokens = ["2","1","+","3","*"]

print(solution.evalRPN(tokens))
```

```python
tokens = ["4","13","5","/","+"]

print(solution.evalRPN(tokens))
```
