```python
class Solution(object):
    def rangeBitwiseAnd(self, left, right):
        """
        :type left: int
        :type right: int
        :rtype: int
        """

        shift = 0

        # Step 1: find the common prefix by shifting right
        while left < right:
            left >>= 1
            right >>= 1
            shift += 1

        # Step 2: shift back to the left to fill with zeros
        return left << shift
```

```python
solution = Solution()
```
